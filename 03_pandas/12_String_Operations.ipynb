# str Accessor

The `.str` accessor in Pandas allows you to apply string methods to every value in a Series without using loops.

```python
import pandas as pd
df = pd.DataFrame({
    "Name": [
        "john doe",
        "ALICE SMITH",
        "Rohit Patel"
    ]
})
df
```

```python
df["Name"].str.upper()
```

```python
df["Name"].str.lower()
```

# Case Conversion

Case conversion methods are used to change the letter casing of text, making data consistent and easier to analyze.

```python
import pandas as pd
df = pd.DataFrame({
    "Name": [
        "john doe",
        "ALICE SMITH",
        "rOhIt pAtEl",
        "jaNe"
    ]
})
df
```

```python
# Lower
df["Name"].str.lower()
```

```python
# Upper
df["Name"].str.upper()
```

```python
# Title:- Capitalize the 1st letter of every word
df["Name"].str.title()
```

```python
# Capitalize only the 1st letter of entire string 
df["Name"].str.capitalize()
```

# Removing Extra Spaces

The `strip()`, `lstrip()`, and `rstrip()` methods are used to remove unwanted spaces from the beginning and/or end of strings.

```python
import pandas as pd
df = pd.DataFrame({
    "Name": [
        "  John Doe  ",
        " Alice ",
        "   Rohit Patel",
        "Jane   "
    ]
})
df
```

```python
 # .strip() :-Removes spaces from both the beginning and the end of a string.
df["Name"].str.strip()
```

```python
#.rstrip():- Removes spaces only from the right (end).
df["Name"].str.rstrip()
```

```python
# .lstrip() :- Removes spaces only from the left (beginning).
df["Name"].str.lstrip()
```

```python
df["Name"] = df["Name"].str.strip().str.title()
df
```

# Searching Text

The `contains()`, `startswith()`, and `endswith()` methods are used to search for specific text patterns in a Pandas Series.

```python
import pandas as pd
df = pd.DataFrame({
    "Name": [
        "John Doe",
        "Alice Smith",
        "Rohit Patel",
        "Jane Doe"
    ],
    "Email": [
        "john@gmail.com",
        "alice@yahoo.com",
        "rohit@gmail.com",
        "jane@outlook.com"
    ]
})
df
```

```python
# .contains() :- Checks whether a string contains a given pattern.
df["Email"].str.contains("gmail")
```

```python
df[df["Email"].str.contains("gmail")]
```

```python
# .startswith():- Checks whether a string starts with a given text.
df["Email"].str.startswith("john")
```

```python
df["Name"].str.startswith("J")
```

```python
# .endswith():- Checks whether a string ends with a given text.
df["Email"].str.endswith(".com")
```

```python
df["Name"].str.endswith("Doe")
```

# split()

The `split()` method splits each string into multiple parts based on a specified separator.

```python
import pandas as pd
df = pd.DataFrame({
    "Name": [
        "John Doe",
        "Alice Smith",
        "Rohit Patel",
        "Jane Brown"
    ],
    "Email": [
        "john@gmail.com",
        "alice@yahoo.com",
        "rohit@gmail.com",
        "jane@outlook.com"
    ]
})
df
```

```python
df["Name"].str.split()
```

```python
df["Email"].str.split("@")
```

```python
df["Name"].str.split(expand=True)  #expand=True tells Pandas:
# "Instead of returning a list, create separate columns."
```

```python
df[["First_Name","Last_Name"]]=df["Name"].str.split(expand=True)
df
```

```python
df["Domain"] = df["Email"].str.split("@").str[1] #.str[1] selects the second element (the domain).
df
```

# replace()

The `replace()` method replaces specified text in each string with new text.

```python
import pandas as pd
df = pd.DataFrame({
    "Name": [
        "John Doe",
        "Alice Smith",
        "Rohit Patel",
        "Jane Doe"
    ],
    "Gender": [
        "Male",
        "Female",
        "Male",
        "Female"
    ],
    "Email": [
        "john@gmail.com",
        "alice@yahoo.com",
        "rohit@gmail.com",
        "jane@outlook.com"
    ]
})
df
```

```python
df["Gender"].str.replace("Male","M")
```

```python
df["Gender"].str.replace("Female","F")
```

```python
df["Gender"] = (
    df["Gender"]
      .str.replace("Male", "M")
      .str.replace("Female", "F")
)
df
```

```python
df["Email"].str.replace("gmail", "googlemail")
```

```python
df["Name"].str.replace(" ", "_")
```

```python
df["Gender"].replace({     #Series Replacement
    "Male": "M",
    "Female": "F"
})
#For entire-value replacements, Series.replace() is cleaner than chaining multiple .str.replace() calls.
```

# extract()

The `extract()` method extracts text from strings using a regular expression (regex) and returns the matched parts.

```python
import pandas as pd
df = pd.DataFrame({
    "Employee": [
        "EMP1001",
        "EMP2050",
        "EMP9876",
        "EMP1234"
    ],
    "Email": [
        "john@gmail.com",
        "alice@yahoo.com",
        "rohit@gmail.com",
        "jane@outlook.com"
    ]
})
df
```

```python
df["Employee"].str.extract(r"(\d+)")
```

```python
df["Employee"].str.extract(r"([A-Za-z]+)")
```

```python
df["Email"].str.extract(r"(.+)@")
```

```python
df["Email"].str.extract(r"@(.+)")
```

```python
import pandas as pd
customers = pd.DataFrame({
    "Name": [
        "  john doe  ",
        "ALICE SMITH",
        "rOhIt pAtEl",
        " jane brown "
    ],
    "Email": [
        "JOHN@GMAIL.COM",
        " alice@yahoo.com ",
        "ROHIT@GMAIL.COM",
        "jane@OUTLOOK.COM"
    ],
    "Gender": [
        "Male",
        "Female",
        "Male",
        "Female"
    ]
})
customers
```

```python
customers["Name"] = (
    customers["Name"]
        .str.strip()
        .str.title()
)

customers["Email"] = (
    customers["Email"]
        .str.strip()
        .str.lower()
)

customers["Gender"] = customers["Gender"].replace({
    "Male": "M",
    "Female": "F"
})

customers["Domain"] = customers["Email"].str.split("@").str[1]
customers
```
