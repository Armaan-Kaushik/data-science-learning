# Apply Functions

Apply functions allow you to perform custom operations on Pandas data. They are useful when built-in methods are not enough and you need to apply your own logic to values, rows, or columns.

df["Column"].map(mapping)

````python
import pandas as pd
df = pd.DataFrame({
    "Employee": ["John", "Alice", "Rohit", "Jane"],
    "Department": ["IT", "HR", "Finance", "IT"],
    "Salary": [50000, 60000, 45000, 70000]
})
df
````

# map()

The `map()` method is used to transform or replace values in a single Pandas Series using a mapping (such as a dictionary, function, or another Series). It applies the specified mapping to each value individually and returns a new Series.

````python
import pandas as pd

df = pd.DataFrame({
    "Employee": ["John", "Alice", "Rohit", "Jane"],
    "Department": ["IT", "HR", "Finance", "IT"],
    "Gender": ["M", "F", "M", "F"]
})

df
````

````python
dept_map={
    "IT":"Information Technology",
    "HR":"Human Resources",
    "Finance":"Finance Department"
}
df["Department"]=df["Department"].map(dept_map)
df
````

````python
gender_map = {
    "M": "Male",
    "F": "Female"
}
df["Gender"].map(gender_map)
````

````python
df["Employee"].map(len)  #map() applies the len() function to every value in the column.
````

````python
df = pd.DataFrame({
    "Department": [
        "IT",
        "HR",
        "Finance",
        "Marketing"
    ]
})
````

````python
dept_map = {
    "IT": "Information Technology",
    "HR": "Human Resources",
    "Finance": "Finance Department"
}

df["Department"].map(dept_map)
````

````python
df["Department"] = (
    df["Department"]
    .map(dept_map)
    .fillna("Other")
)
````

# apply()

The `apply()` method is used to apply a custom function to each value in a Series or to each row or column in a DataFrame. It is useful when built-in Pandas functions cannot perform the required operation.

df["Column"].apply(function)

````python
import pandas as pd
df = pd.DataFrame({
    "Employee": ["John", "Alice", "Rohit", "Jane"],
    "Salary": [50000, 60000, 45000, 70000]
})
df
````

````python
#Give every employee a 10% salary hike.
def increase_salary(salary):
    return salary * 1.10
df["Salary"].apply(increase_salary)
````

````python
df["New_Salary"] = df["Salary"].apply(increase_salary)
df
````

````python
#Create Salary Category

def salary_category(salary):
    if salary >= 60000:
        return "High"
    else:
        return "Normal"
df["Salary"].apply(salary_category)
````

````python
df["Category"] = df["Salary"].apply(salary_category)

df
````

````python
df["Employee"].apply(len)
````

# apply() on DataFrame

The `apply()` method can be used on an entire DataFrame to apply a function to each row or each column. The `axis` parameter determines whether the function operates column-wise (`axis=0`) or row-wise (`axis=1`).

df.apply(function, axis=?)

````python
import pandas as pd
df = pd.DataFrame({
    "Employee": ["John", "Alice", "Rohit", "Jane"],
    "Sales": [1200, 800, 1500, 900],
    "Target": [1000, 1000, 1200, 1000]
})
df
````

````python
df[["Sales", "Target"]].apply(max)
````

````python
def check_target(row):
    if row["Sales"] >= row["Target"]:
        return "Achieved"
    else:
        return "Missed"
df["Status"] = df.apply(check_target, axis=1)
df
````

````python
def total(row):
    return row["Sales"] + row["Target"]
df["Total"] = df.apply(total, axis=1)
df
````

# lambda

A `lambda` function is a small anonymous (unnamed) function used to write simple operations in a single line. It is commonly used with methods like `apply()` and `map()` for quick transformations.

def function(x):

    return x * 2

↓

lambda x: x * 2

````python
import pandas as pd
df = pd.DataFrame({
    "Employee": ["John", "Alice", "Rohit", "Jane"],
    "Salary": [50000, 60000, 45000, 70000]
})
df
````

````python
df["Salary"].apply(lambda salary: salary * 1.10)
````

````python
df["Salary"].apply(
    lambda salary: "High" if salary >= 60000 else "Normal"
)
````

````python
df["Employee"].apply(lambda name: len(name))
````

````python
df["Employee"].map(lambda name: name.upper())
````

````python
df = pd.DataFrame({
    "Employee": ["John", "Alice", "Rohit", "Jane"],
    "Sales": [1200, 800, 1500, 900],
    "Target": [1000, 1000, 1200, 1000]
})
````

````python
df["Status"] = df.apply(
    lambda row: "Achieved"
    if row["Sales"] >= row["Target"]
    else "Missed",
    axis=1
)
df
````

# When to Use map(), apply(), and lambda

`map()`, `apply()`, and `lambda` are used for transforming data in Pandas. Choosing the correct one depends on whether you're working with a single column or multiple columns, and whether the transformation is simple or requires custom logic.

# Notebook 14 Summary

| Topic | Purpose |
|--------|---------|
| `map()` | Replace or transform values in a single Series (column) using a dictionary, function, or another Series. |
| `apply()` (Series) | Apply a custom or built-in function to each value in a single column. |
| `apply()` (DataFrame) | Apply a function to each row (`axis=1`) or each column (`axis=0`) of a DataFrame. |
| `lambda` | Write short, anonymous one-line functions, commonly used with `map()` and `apply()`. |
| `axis=0` | Apply a function to each column of a DataFrame. |
| `axis=1` | Apply a function to each row of a DataFrame. |

---

# When to Use Each

| Situation | Use |
|-----------|-----|
| Replace values using a dictionary | `map()` |
| Apply a function to one column | `apply()` |
| Compare multiple columns | `apply(axis=1)` |
| Perform operations on each column | `apply(axis=0)` |
| Write a short one-line function | `lambda` |
| Write complex or reusable logic | `def` + `apply()` |

---

# Quick Comparison

| Feature | `map()` | `apply()` | `lambda` |
|---------|---------|-----------|----------|
| Works on Series | ✅ | ✅ | Used with `map()`/`apply()` |
| Works on DataFrame | ❌ | ✅ | ✅ (inside `apply`) |
| Dictionary Mapping | ✅ | ❌ | ❌ |
| Custom Functions | ✅ | ✅ | ✅ |
| Multiple Columns | ❌ | ✅ (`axis=1`) | ✅ (`axis=1`) |
| One-Line Logic | ❌ | ✅ | ✅ Best Choice |
| Complex Logic | ❌ | ✅ | ❌ |
