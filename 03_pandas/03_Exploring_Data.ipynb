```python
import pandas as pd

employees = {
    "Employee_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Priya", "Rahul", "Sneha", "Vikas", "Neha"],
    "Department": ["HR", "IT", "Finance", "IT", "Sales", "HR"],
    "Age": [25, 30, 28, 35, 40, 27],
    "Salary": [35000, 60000, 50000, 75000, 45000, 38000]
}

df = pd.DataFrame(employees)

df
```

# 📌 head()

> **Definition:** `head()` is a Pandas method used to display the first few rows of a DataFrame. By default, it returns the first **5 rows**, allowing us to quickly inspect the dataset.

```python
df.head()
```

```python
df.head(3)
```

# 📌 tail()

> **Definition:** `tail()` is a Pandas method used to display the last few rows of a DataFrame. By default, it returns the last **5 rows**, allowing us to inspect the end of the dataset.

```python
df.tail()
```

```python
df.tail(3)
```

# 📌 info()

> **Definition:** `info()` is a Pandas method that displays a concise summary of a DataFrame, including the number of rows, column names, data types, non-null values, and memory usage.

```python
df.info()
```

# 📌 describe()

> **Definition:** `describe()` is a Pandas method that generates descriptive statistics for numerical columns in a DataFrame, such as count, mean, standard deviation, minimum, maximum, and quartile values.

```python
df.describe()
```

```python
#For Single Element
print("Average Age :", df["Age"].mean())  
print("Maximum Age :", df["Age"].max())
print("Minimum Age :", df["Age"].min())
```

# 📌 shape

> **Definition:** `shape` is a Pandas DataFrame attribute that returns the dimensions of a DataFrame as a tuple in the form **(number of rows, number of columns)**.

```python
print(df.shape)
print(df.shape[0])
print(df.shape[1])
```

# 📌 columns

> **Definition:** `columns` is a Pandas DataFrame attribute that returns the names (labels) of all columns in a DataFrame as an Index object.

```python
print(df.columns)
```

```python
print(df.columns.tolist())
```

# 📌 index

> **Definition:** `index` is a Pandas DataFrame attribute that returns the row labels (indexes) of a DataFrame as an Index object. By default, Pandas assigns indexes starting from **0**.

```python
print(df.index)
```

```python
print(df.index.tolist())
```

```python
df.index[-1]
```

# 📌 dtypes

> **Definition:** `dtypes` is a Pandas DataFrame attribute that displays the data type of each column in a DataFrame. It helps us identify whether a column contains integers, floating-point numbers, text, boolean values, dates, or other data types.

```python
print(df.dtypes)
```

```python
print(df[["Name","Age"]].dtypes)
```
