# Advanced Data Cleaning

Advanced Data Cleaning is the process of identifying and correcting issues in a dataset such as duplicate records, missing or incorrect values, inconsistent naming, and outliers to improve data quality before analysis.

# Duplicates

Duplicates are rows in a dataset that contain identical or repeated information. Identifying and removing duplicate records helps improve data quality and prevents incorrect analysis.

```python
import pandas as pd
df = pd.DataFrame({
    "Customer": ["John", "Alice", "John", "Bob", "Alice"],
    "City": ["Mumbai", "Pune", "Mumbai", "Delhi", "Pune"],
    "Sales": [1200, 2500, 1200, 1800, 2500]
})

df
```

```python
df.duplicated()
```

```python
df[df.duplicated()]
```

```python
df.duplicated().sum()
```

```python
# Duplicate Based on Specific Columns
df.duplicated(subset=["Customer"])
```

```python
df.duplicated(subset=["Customer","City"])
```

# drop_duplicates()

The `drop_duplicates()` method is used to remove duplicate rows from a DataFrame. By default, it keeps the first occurrence of each duplicate and removes the remaining ones.

```python
import pandas as pd

df = pd.DataFrame({
    "Customer": ["John", "Alice", "John", "Bob", "Alice"],
    "City": ["Mumbai", "Pune", "Mumbai", "Delhi", "Pune"],
    "Sales": [1200, 2500, 1200, 1800, 2500]
})

df
```

```python
df.duplicated()
```

```python
df.drop_duplicates()
```

```python
df
```

```python
df.drop_duplicates(keep="first",ignore_index=True)
```

```python
df.drop_duplicates(subset=["Customer"], keep="last",ignore_index=True)
```

```python
df.drop_duplicates(keep=False)
```

```python
df.drop_duplicates(subset=["Customer"],ignore_index=True)
```

# Outliers

Outliers are data points that are significantly different from the rest of the observations in a dataset. They can occur due to data entry errors, measurement errors, or genuine extreme values, and may affect statistical analysis and machine learning models.

```python
import pandas as pd
df = pd.DataFrame({
    "Salary": [45000, 50000, 48000, 52000, 47000, 49000, 500000]
})
df
```

```python
Q1 = df["Salary"].quantile(0.25)
Q3 = df["Salary"].quantile(0.75)
print(Q1)
print(Q3)
IQR = Q3 - Q1
print(IQR)
```

```python
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
print(lower)
print(upper)
```

```python
df[
    (df["Salary"] < lower) |
    (df["Salary"] > upper)
]
```

# rename()

The `rename()` method is used to change the names of columns or index labels in a Pandas DataFrame without modifying the actual data.

df.rename(columns={"old_name": "new_name"})

```python
import pandas as pd
df = pd.DataFrame({
    "emp_id": [101, 102, 103],
    "emp_name": ["John", "Alice", "Rohit"],
    "sal": [50000, 60000, 45000]
})
df
```

```python
df.rename(columns={"sal":"Salary"})
```

```python
df  #Original Dataframe is not changed
```

```python
df.rename(columns={
    "emp_id": "Employee_ID",
    "emp_name": "Employee_Name",
    "sal": "Salary"
})
```

```python
df.rename(index={
    0:"A",
    1:"B",
    2:"C"
})
```

```python
df.rename(
    index={0: "First"},
    columns={"sal": "Salary"}
)
```

# replace()

The `replace()` method is used to replace specific values in a Pandas Series or DataFrame with new values. It is commonly used to correct incorrect, inconsistent, or unwanted data.

df.replace(old_value, new_value)

```python
import pandas as pd
df = pd.DataFrame({
    "Name": ["John", "Alice", "Rohit", "Jane"],
    "City": ["pune", "Mumbai", "PUNE", "pune"],
    "Gender": ["M", "F", "M", "F"]
})
df
```

```python
df.replace("pune", "Pune")
```

```python
df.replace({
    "M": "Male",
    "F": "Female"
})
```

```python
df
```

```python
df.replace({
    "pune": "Pune",
    "PUNE": "Pune",
    "M": "Male",
    "F": "Female"
})
```

```python
df
```

```python
df["Gender"] = df["Gender"].replace({
    "M":"Male",
    "F":"Female"
})
```

```python
df
```

```python
df["City"] = df["City"].replace({
    "pune": "Pune",
    "PUNE": "Pune"
})
```

```python
df
```

# Real-World Data Cleaning Workflow

In real-world projects, datasets often contain duplicate records, inconsistent values, poorly named columns, and outliers. Before performing any analysis, these issues should be identified and corrected to ensure accurate results.

This workflow demonstrates how to clean a messy dataset using common Pandas techniques.

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({
    "cust_name": ["John", "Alice", "John", "Bob", "David", "Jane"],
    "city": ["pune", "Mumbai", "pune", "Delhi", "Unknown", "PUNE"],
    "gender": ["M", "F", "M", "M", "F", "F"],
    "sales": [1200, 2500, 1200, 1800, 500000, 2200]
})

df
```

```python
df = df.drop_duplicates(ignore_index=True)
```

```python
df = df.replace({
    "pune": "Pune",
    "PUNE": "Pune",
    "Unknown": np.nan,
    "M": "Male",
    "F": "Female"
})
df
```

```python
df = df.rename(columns={
    "cust_name": "Customer_Name",
    "city": "City",
    "gender": "Gender",
    "sales": "Sales"
})
df
```

```python
Q1 = df["Sales"].quantile(0.25)
Q3 = df["Sales"].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
```

```python
df = df[
    (df["Sales"] >= lower) &
    (df["Sales"] <= upper)
]
df
```
