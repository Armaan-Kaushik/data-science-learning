```python
import pandas as pd
employees = pd.DataFrame({
    "Emp_ID": [101,102,103,104,105,106,107,108],
    "Name": ["Amit","Neha","Rahul","Sneha","Karan","Priya","Vikas","Anjali"],
    "Department": ["HR","IT","Finance","IT","HR","Finance","IT","HR"],
    "Gender": ["Male","Female","Male","Female","Male","Female","Male","Female"],
    "Experience": [2,5,4,8,3,6,2,7],
    "Salary": [45000,70000,65000,85000,48000,72000,60000,55000]
})
employees
```

# Definition: groupby()

`groupby()` is a Pandas method used to divide data into groups based on one or more columns so that calculations can be performed separately on each group.

```python
# Syntax: df.groupby("column_name")
#         df.groupby(["column1", "column2"])
```

```python
employees.groupby("Department")
```

# Aggregation using mean()

Aggregation means combining multiple values into a single summarized value.

```python
# Avg Salary of each Department
employees.groupby("Department")["Salary"].mean()
```

```python
#Avg Experience of each Department
employees.groupby("Department")["Experience"].mean()
```

```python
employees.groupby("Department")[["Salary","Experience"]].mean()
```

# Aggregation Functions with groupby()

Aggregation functions summarize the data in each group into a single value such as sum, average, count, minimum, or maximum.

```python
# Sum
employees.groupby("Department")["Salary"].mean()
```

```python
# Count
employees.groupby("Department")["Salary"].count()
# count() ignores missing values.
```

```python
# Size  :-  Find the total number of rows in each department.
employees.groupby("Department").size()
# size() counts all rows.
```

```python
# min()
employees.groupby("Department")["Salary"].min()
```

```python
# Max()
employees.groupby("Department")["Salary"].max()
```

```python
# Median
employees.groupby("Department")["Salary"].median()
```

```python
# std()  #To calculate standard Deviation
employees.groupby("Department")["Salary"].std()
```

```python
# Var  #To calculate variance
employees.groupby("Department")["Salary"].var()
```

```python
# First To show first value of every group
employees.groupby("Department").first()
```

```python
# last()    To show last value of every group
employees.groupby("Department").last()
```

```python
#  nunique()    Count unique values.
employees.groupby("Department")["Gender"].nunique()
```

# Multiple Aggregations using agg()

`agg()` (aggregate) allows us to perform one or more aggregation functions on grouped data in a single operation.

```python
#  df.groupby("column").agg(functions)
             
            #or

# df.groupby("column").agg({
#   "column1": [...],
#  "column2": [...]
#  })
```

```python
employees.groupby("Department")["Salary"].agg(
    ["mean", "max", "min"]
)
```

```python
employees.groupby("Department")["Salary"].agg(
    ["count", "sum", "mean", "median", "min", "max"]
)
```

```python
employees.groupby("Department").agg({
    "Salary": ["mean", "max"],
    "Experience": "mean"
})
```

```python
employees.groupby("Department").agg(
    Average_Salary=("Salary", "mean"),
    Highest_Salary=("Salary", "max"),
    Lowest_Salary=("Salary", "min"),
    Avg_Experience=("Experience", "mean")
)
```

```python
employees.groupby("Department").agg(
    Total_Salary=("Salary", "sum"),
    Average_Salary=("Salary", "mean"),
    Highest_Salary=("Salary", "max"),
    Employees=("Emp_ID", "count")
)
```

# Grouping by Multiple Columns

Grouping by multiple columns allows us to divide the data into groups based on combinations of two or more columns, enabling more detailed analysis.

```python
employees.groupby(["Department", "Gender"])
```

```python
employees.groupby(["Department", "Gender"])["Salary"].mean()
```

```python
employees.groupby(
    ["Department", "Gender"]
)["Salary"].agg(["mean", "max"])
```

```python
employees.groupby(
    ["Department", "Gender"]
).agg(
    Average_Salary=("Salary", "mean"),
    Highest_Salary=("Salary", "max"),
    Employees=("Emp_ID", "count")
)
```

```python
employees.groupby(
    ["Department", "Gender"]
).agg(
    Average_Salary=("Salary", "mean"),
    Highest_Salary=("Salary", "max"),
    Employees=("Emp_ID", "count")
).reset_index()
```

```python
employees.groupby(
    ["Department", "Gender"]
)["Salary"].mean().reset_index()  #reset_index() converts the index levels back into normal columns.
```
