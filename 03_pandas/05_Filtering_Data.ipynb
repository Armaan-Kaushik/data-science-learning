## Basic Condition Filtering

**Definition:**
Basic condition filtering selects only those rows that satisfy a given condition.

**Syntax:**
dataframe[condition]

```python
import pandas as pd

employees = pd.DataFrame({
    "Employee": ["Amit", "Neha", "Rahul", "Priya", "Karan"],
    "Department": ["HR", "IT", "Sales", "IT", "Finance"],
    "Salary": [45000, 70000, 55000, 80000, 60000]
})

employees
```

```python
high_salary=employees[employees["Salary"]> 60000]
high_salary
```

```python
print(employees[employees["Salary"] != 60000])
```

```python
print(employees[employees["Department"]=="IT"])
```

## Multiple Condition Filtering

**Definition:**
Multiple condition filtering selects rows that satisfy two or more conditions at the same time.

**Syntax:**
dataframe[(condition1) & (condition2)]

dataframe[(condition1) | (condition2)]

dataframe[~condition]

```python
import pandas as pd

employees = pd.DataFrame({
    "Employee": ["Amit", "Neha", "Rahul", "Priya", "Karan", "Sneha"],
    "Department": ["HR", "IT", "Sales", "IT", "Finance", "IT"],
    "Salary": [45000, 70000, 55000, 80000, 60000, 50000],
    "Experience": [2, 5, 3, 7, 4, 1]
})

employees
```

```python
print(employees[(employees["Department"]=="IT")&(employees["Salary"]>60000)])
```

```python
print(employees[(employees["Department"]=="HR")|(employees["Department"]=="Finance")])
```

```python
print(employees[~(employees["Department"]=="IT")])
```

## isin()

**Definition:**
Checks whether each value in a column exists in a given list of values.

**Syntax:**
dataframe[dataframe["column"].isin([value1, value2, ...])]

```python
import pandas as pd

employees = pd.DataFrame({
    "Employee": ["Amit", "Neha", "Rahul", "Priya", "Karan", "Sneha"],
    "Department": ["HR", "IT", "Sales", "IT", "Finance", "IT"],
    "Salary": [45000, 70000, 55000, 80000, 60000, 50000]
})

employees
```

```python
print(employees[employees["Department"].isin(["IT","HR"])])
```

```python
print(employees[employees["Salary"].isin([45000,70000])])
```

```python
print(employees[employees["Employee"].isin(["Amit","Priya","Sneha"])])
```

```python
print(employees[~(employees["Department"].isin(["IT"]))])
```

## between()

**Definition:**
Checks whether values lie between two specified values.

**Syntax:**
dataframe[dataframe["column"].between(start_value, end_value)]

```python
import pandas as pd

employees = pd.DataFrame({
    "Employee": ["Amit", "Neha", "Rahul", "Priya", "Karan", "Sneha"],
    "Department": ["HR", "IT", "Sales", "IT", "Finance", "IT"],
    "Salary": [45000, 70000, 55000, 80000, 60000, 50000]
})

employees
```

```python
print(employees[employees["Salary"].between(50000,70000)])
```

```python
employees["Experience"]=[2,5,3,7,4,1]
print(employees)
```

```python
print(employees[employees["Experience"].between(2,5)])
```

```python
print(employees[employees["Employee"].between("K", "R")])
```

## query()

**Definition:**
Filters rows by evaluating a condition written as a string.

**Syntax:**
dataframe.query("condition")

```python
import pandas as pd

employees = pd.DataFrame({
    "Employee": ["Amit", "Neha", "Rahul", "Priya", "Karan", "Sneha"],
    "Department": ["HR", "IT", "Sales", "IT", "Finance", "IT"],
    "Salary": [45000, 70000, 55000, 80000, 60000, 50000],
    "Experience": [2, 5, 3, 7, 4, 1]
})

employees
```

```python
print(employees.query("Salary >60000"))
```

```python
print(employees.query("Department == 'IT' and Salary > 60000"))
```

```python
print(employees.query("Salary >= 55000 and Experience > 3"))
```

```python
minimum_salary = 60000
print(employees.query("Salary > @minimum_salary"))
```

The @ symbol tells query() to use the value of a Python variable.

Without @, Pandas will search for a column named minimum_salary.
