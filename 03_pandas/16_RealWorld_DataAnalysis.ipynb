# Real World Data Analysis

Real World Data Analysis involves exploring, cleaning, transforming, and analyzing business data to answer practical questions and support decision-making. Instead of learning individual functions, the focus is on solving real business problems using a complete data analysis workflow.

```python
import pandas as pd
df = pd.DataFrame({
    "Order_ID": [101,102,103,104,105,106,107,108,109,110],
    "Customer": ["John","Alice","Bob","John","David","Alice","Eva","Frank","Grace","Henry"],
    "Region": ["North","South","East","North","West","South","East","West","North","South"],
    "Category": ["Electronics","Furniture","Electronics","Clothing","Furniture",
                 "Electronics","Clothing","Furniture","Electronics","Clothing"],
    "Product": ["Laptop","Chair","Phone","T-Shirt","Table",
                "Headphones","Jeans","Sofa","Tablet","Jacket"],
    "Salesperson": ["Amit","Neha","Amit","Raj","Neha",
                    "Raj","Amit","Neha","Raj","Amit"],
    "Quantity": [2,4,3,5,2,6,4,1,3,2],
    "Price": [50000,7000,30000,1200,15000,2500,1800,40000,28000,3500],
    "Discount": [10,5,15,0,8,5,0,20,10,5],
    "Order_Date": pd.to_datetime([
        "2024-01-15","2024-01-20","2024-02-10","2024-02-18",
        "2024-03-05","2024-03-12","2024-04-01","2024-04-18",
        "2024-05-09","2024-05-25"
    ])
})
df
```

# Business Question 1

## Which products generate the highest revenue?

Businesses analyze product revenue to identify their best-selling and highest-earning products. This helps management make decisions about inventory, marketing, pricing, and future investments.

```python
df.groupby("Product")
```

```python
df.groupby("Product")["Revenue"].sum()
```

# Business Question 5

## How does revenue change over time?

Businesses monitor monthly sales trends to identify seasonality, growth patterns, and periods of high or low performance. This analysis helps in forecasting, inventory planning, and marketing decisions.

```python
df["Month"] = df["Order_Date"].dt.month
df[["Order_Date", "Month"]]
```

```python
df["Month_Name"] = df["Order_Date"].dt.month_name()
df[["Order_Date", "Month_Name"]]
```

```python
df.groupby("Month_Name")["Revenue"].sum()
```

```python
monthly_revenue = (
    df.groupby(["Month", "Month_Name"])["Revenue"]
      .sum()
      .reset_index()
)
monthly_revenue
```

```python
monthly_revenue = (
    df.groupby(["Month", "Month_Name"])["Revenue"]
      .sum()
      .reset_index()
      .drop(columns="Month")
)
monthly_revenue
```

```python
monthly_revenue.sort_values(
    by="Revenue",
    ascending=False
)
```

# Management Summary Report

Managers do not usually look at raw data. Instead, they prefer concise summary reports that highlight key performance indicators (KPIs). Using `groupby()`, `agg()`, and `pivot_table()`, we can create reports that support business decision-making.

```python
region_report = (
    df.groupby("Region")
      .agg(
          Total_Revenue=("Revenue", "sum"),
          Total_Orders=("Order_ID", "count"),
          Average_Discount=("Discount", "mean")
      )
      .sort_values("Total_Revenue", ascending=False)
)

region_report
```

```python
category_report = (
    df.groupby("Category")
      .agg(
          Total_Revenue=("Revenue", "sum"),
          Orders=("Order_ID", "count"),
          Average_Discount=("Discount", "mean")
      )
      .sort_values("Total_Revenue", ascending=False)
)
category_report
```

```python
salesperson_report = (
    df.groupby("Salesperson")
      .agg(
          Total_Revenue=("Revenue", "sum"),
          Orders=("Order_ID", "count"),
          Average_Revenue=("Revenue", "mean")
      )
      .sort_values("Total_Revenue", ascending=False)
)
salesperson_report
```

```python
customer_report = (
    df.groupby("Customer")
      .agg(
          Total_Revenue=("Revenue", "sum"),
          Orders=("Order_ID", "count")
      )
      .sort_values("Total_Revenue", ascending=False)
)
customer_report
```

```python
pivot = pd.pivot_table(
    df,
    values="Revenue",
    index="Region",
    columns="Category",
    aggfunc="sum",
    fill_value=0
)
pivot
```
