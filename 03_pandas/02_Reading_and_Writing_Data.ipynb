## Exploring the Data After Reading

After loading a dataset into a DataFrame, the first step is to inspect its contents. Pandas provides several built-in functions to quickly understand the structure, size, data types, and summary of the dataset before performing any analysis.

## head()

The `head()` method displays the first few rows of a DataFrame. By default, it returns the first **5 rows**, making it useful for quickly checking whether the data has been loaded correctly.

```python
import pandas as pd
df =pd.read_csv("employee_data.csv")
print(df.head())
```

```python
print(df.head(3))
```

## tail()

The `tail()` method displays the last few rows of a DataFrame. By default, it returns the last **5 rows**.

```python
print(df.tail())
```

```python
print(df.tail(2))
```

## Understanding info(), describe(), and shape()

After checking the first and last few rows of a dataset, the next step is to understand its overall structure. Pandas provides the `info()`, `describe()`, and `shape` methods to quickly inspect the dataset before performing any analysis.

## info()

The `info()` method provides a concise summary of a DataFrame. It displays the number of rows, columns, data types, non-null values, memory usage, and other useful information.

```python
print(df.info())
```

## describe()

The `describe()` method generates descriptive statistics for numerical columns in a DataFrame. It helps summarize the distribution of the data.

```python
print(df.describe(),"\n")
print(df.describe(include="all"))  #also summarizes text columns.
```

## shape

The `shape` attribute returns the dimensions of the DataFrame as a tuple in the form `(rows, columns)`.

```python
print(df.shape)
print(df.shape[0])
print(df.shape[1])
```

The `index_col` parameter is used to specify which column should be used as the row index of the DataFrame instead of the default numeric index.

## usecols

The `usecols` parameter allows us to read only the required columns from a CSV file instead of loading the entire dataset. This improves performance and reduces memory usage, especially when working with large files.

```python
import pandas as pd
df=pd.read_csv(
    "employee_data.csv",
    index_col="EmployeeID"   
)
print(df)
```

```python
df = pd.read_csv(
    "employee_data.csv",
    usecols=[0, 2]
)

print(df)
```

## nrows Parameter

## dtype Parameter

The `dtype` parameter is used to explicitly specify the data type of one or more columns while reading a CSV file. It helps prevent incorrect data type inference, reduces memory usage, and ensures consistent data processing.

## parse_dates Parameter

The `parse_dates` parameter is used to automatically convert one or more columns into Pandas datetime objects while reading a CSV file. This makes it easier to perform date-based filtering, sorting, calculations, and time series analysis.

```python
import pandas as pd

sales_data = {
    "OrderID": [1001, 1002, 1003],
    "Customer": ["Rahul", "Priya", "Amit"],
    "OrderDate": ["2026-07-10", "2026-07-12", "2026-07-15"],
    "Amount": [5000, 7200, 6800]
}

df = pd.DataFrame(sales_data)
df.to_csv("sales_data.csv", index=False)
print("sales_data.csv created successfully!")
```

```python
import pandas as pd
df = pd.read_csv(
    "sales_data.csv",
    parse_dates=["OrderDate"]
)
print(df.dtypes)
```

```python
print(df["OrderDate"].dt.year)
print(df["OrderDate"].dt.month)
print(df["OrderDate"].dt.day)
```

```python
print(df["OrderDate"].dt.day_name())
```

## Reading Excel Files using read_excel()

While CSV files are widely used for data exchange, Excel files are the most common format in business environments. Pandas provides the `read_excel()` function to read data from Excel (.xlsx or .xls) files efficiently.

```python
import pandas as pd

employee_data = {
    "EmployeeID": [101, 102, 103, 104, 105],
    "Name": ["Rahul", "Priya", "Amit", "Neha", "Karan"],
    "Department": ["IT", "HR", "Finance", "Marketing", "IT"],
    "Salary": [65000, 58000, 72000, 61000, 69000]
}
df = pd.DataFrame(employee_data)
df.to_excel("employee_data.xlsx", index=False)
print("employee_data.xlsx created successfully!")
```

```python
import pandas as pd

df = pd.read_excel("employee_data.xlsx")
print(df)
```

## sheet_name Parameter

The `sheet_name` parameter is used to specify which worksheet should be read from an Excel file. Since an Excel workbook can contain multiple sheets, this parameter allows us to read a single sheet, multiple sheets, or all sheets.

```python
import pandas as pd

employees = pd.DataFrame({
    "EmployeeID": [101, 102, 103],
    "Name": ["Rahul", "Priya", "Amit"]
})

departments = pd.DataFrame({
    "EmployeeID": [101, 102, 103],
    "Department": ["IT", "HR", "Finance"]
})

salaries = pd.DataFrame({
    "EmployeeID": [101, 102, 103],
    "Salary": [65000, 58000, 72000]
})

with pd.ExcelWriter("company_data.xlsx") as writer:
    employees.to_excel(writer, sheet_name="Employees", index=False)
    departments.to_excel(writer, sheet_name="Departments", index=False)
    salaries.to_excel(writer, sheet_name="Salaries", index=False)

print("company_data.xlsx created successfully!")
```

```python
df = pd.read_excel("company_data.xlsx")
print(df)
```

```python
df = pd.read_excel(
    "company_data.xlsx",
    sheet_name="Departments"
)
print(df)
```

```python
df = pd.read_excel(
    "company_data.xlsx",
    sheet_name=2
)
print(df)
```

```python
data = pd.read_excel(
    "company_data.xlsx",
    sheet_name=["Employees","Salaries"]
)

print(data)
```

## Common read_excel() Parameters

Most parameters of `read_excel()` work exactly like `read_csv()`. The only major difference is that Excel files contain worksheets instead of comma-separated text. Therefore, once you know `read_csv()`, learning `read_excel()` becomes very easy.

```python
df = pd.read_excel(
    "company_data.xlsx",
    sheet_name="Employees",
    header=0
)
print(df)
```

```python
df = pd.read_excel(
    "company_data.xlsx",
    sheet_name="Employees",
    header=0,
    names=["ID", "Employee_Name"]
)
print(df)
```

```python
df = pd.read_excel(
    "company_data.xlsx",
    sheet_name="Employees",
    usecols=["EmployeeID", "Name"]   #usecols=[0,1]
)
print(df)
```

```python
df = pd.read_excel(
    "company_data.xlsx",
    sheet_name="Employees",
    skiprows=2
)
print(df)
```

```python
df = pd.read_excel(
    "company_data.xlsx",
    sheet_name="Employees",
    nrows=2
)
print(df)
```

```python
df = pd.read_excel(
    "company_data.xlsx",
    sheet_name="Employees",
    dtype={
        "EmployeeID": str
    }
)
print(df.dtypes)
```

## to_excel()

The `to_excel()` method is used to export a Pandas DataFrame to an Excel (.xlsx) file. It is commonly used to share reports, dashboards, and processed datasets with business users.

```python
import pandas as pd
employee_data = {
    "EmployeeID": [101, 102, 103],
    "Name": ["Rahul", "Priya", "Amit"],
    "Department": ["IT", "HR", "Finance"],
    "Salary": [65000, 58000, 72000]
}
df = pd.DataFrame(employee_data)
```

```python
df.to_excel(
    "employees.xlsx",
    index=False # Used to remove Index by defult it is true so indexing is done automatically
)
print("Excel file created successfully")
```

```python
employees = pd.DataFrame({
    "EmployeeID": [101, 102],
    "Name": ["Rahul", "Priya"]
})

departments = pd.DataFrame({
    "EmployeeID": [101, 102],
    "Department": ["IT", "HR"]
})
```

```python
with pd.ExcelWriter("company_data.xlsx") as writer:
    employees.to_excel(
        writer,
        sheet_name="Employees",
        index=False
    )

    departments.to_excel(
        writer,
        sheet_name="Departments",
        index=False
    )

print("Excel workbook created successfully")
```
