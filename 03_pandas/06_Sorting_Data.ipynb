## sort_values()

**Definition:**
Sorts the rows of a DataFrame based on the values of one or more columns.

**Syntax:**
dataframe.sort_values(by="column_name")

```python
import pandas as pd

employees = pd.DataFrame({
    "Employee": ["Amit", "Neha", "Rahul", "Priya", "Karan"],
    "Department": ["HR", "IT", "Sales", "IT", "Finance"],
    "Salary": [45000, 70000, 55000, 80000, 60000],
    "Experience": [2, 5, 3, 7, 4]
})

employees
```

```python
print(employees.sort_values(by="Salary"))
```

## ascending Parameter

**Definition:**
Controls the sorting order in `sort_values()`. It can sort data in ascending or descending order.

**Syntax:**
dataframe.sort_values(by="column_name", ascending=True)

dataframe.sort_values(by="column_name", ascending=False)

```python
import pandas as pd
employees = pd.DataFrame({
    "Employee": ["Amit", "Neha", "Rahul", "Priya", "Karan"],
    "Department": ["HR", "IT", "Sales", "IT", "Finance"],
    "Salary": [45000, 70000, 55000, 80000, 60000],
    "Experience": [2, 5, 3, 7, 4]
})
employees
```

```python
print(employees.sort_values(by="Salary",ascending=True))
print("\n",employees.sort_values(by="Salary",ascending=False))
```

```python
print(employees.sort_values(by="Employee",ascending=False))
```

```python
print(employees.sort_values(by="Experience",ascending=False))
```

## Sorting by Multiple Columns

**Definition:**
Sorts a DataFrame using two or more columns in a specified order.

**Syntax:**
dataframe.sort_values(by=["column1", "column2"])

dataframe.sort_values(
    by=["column1", "column2"],
    ascending=[True, False]
)

```python
import pandas as pd
employees = pd.DataFrame({
    "Employee": ["Amit", "Neha", "Rahul", "Priya", "Karan", "Sneha"],
    "Department": ["HR", "IT", "Sales", "IT", "Finance", "IT"],
    "Salary": [45000, 70000, 55000, 80000, 60000, 50000],
    "Experience": [2, 5, 3, 7, 4, 1]
})
employees
```

```python
print(employees.sort_values(by=["Department","Salary"]))
```

```python
print(employees.sort_values(
    by=["Department", "Salary"],
    ascending=[True, False]
))
```

```python
print(employees.sort_values(by=["Department","Experience"],ascending=[True,True]))
```

## sort_index()

**Definition:**
Sorts a DataFrame based on its index labels.

**Syntax:**
dataframe.sort_index()

dataframe.sort_index(ascending=False)

```python
import pandas as pd
employees = pd.DataFrame({
    "Employee": ["Amit", "Neha", "Rahul", "Priya", "Karan"],
    "Department": ["HR", "IT", "Sales", "IT", "Finance"],
    "Salary": [45000, 70000, 55000, 80000, 60000]
})
employees
```

```python
sorted_salary=employees.sort_values(by="Salary")
sorted_salary
```
