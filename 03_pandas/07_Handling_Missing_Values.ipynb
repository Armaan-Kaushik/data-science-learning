## isna()

**Definition:**
Checks whether each value in a DataFrame or Series is a missing value (`NaN`).

**Syntax:**
dataframe.isna()

series.isna()

```python
import pandas as pd
import numpy as np

employees = pd.DataFrame({
    "Employee": ["Amit", "Neha", "Rahul", "Priya", "Karan"],
    "Department": ["HR", "IT", np.nan, "IT", "Finance"],
    "Salary": [45000, np.nan, 55000, 80000, np.nan],
    "Experience": [2, 5, 3, np.nan, 4]
})

employees
```

```python
print(employees.isna())
```

```python
print(employees[employees["Salary"].isna()])
```

```python
employees.isna().sum() #Find the total number of missing values in each column.
```

## notna()

**Definition:**
Checks whether each value in a DataFrame or Series is **not** a missing value (`NaN`).

**Syntax:**
dataframe.notna()

series.notna()

```python
import pandas as pd
import numpy as np

employees = pd.DataFrame({
    "Employee": ["Amit", "Neha", "Rahul", "Priya", "Karan"],
    "Department": ["HR", "IT", np.nan, "IT", "Finance"],
    "Salary": [45000, np.nan, 55000, 80000, np.nan],
    "Experience": [2, 5, 3, np.nan, 4]
})

employees
```

```python
print(employees.notna())
```

```python
employees["Salary"].notna()
```

```python
employees[employees["Salary"].notna()]
```

```python
employees.notna().sum()
```

## dropna()

**Definition:**
Removes rows or columns containing missing values (`NaN`).

**Syntax:**
dataframe.dropna()

dataframe.dropna(axis=1)

```python
import pandas as pd
import numpy as np
employees = pd.DataFrame({
    "Employee": ["Amit", "Neha", "Rahul", "Priya", "Karan"],
    "Department": ["HR", "IT", np.nan, "IT", "Finance"],
    "Salary": [45000, np.nan, 55000, 80000, np.nan],
    "Experience": [2, 5, 3, np.nan, 4]
})
employees
```

```python
employees.dropna()
```

```python
employees.dropna(axis=0)  #Remove all columns that contain at least one NaN.
```

```python
employees.dropna(axis=1)
```

```python
employees.dropna(axis=0)  #Remove all rows that contain at least one NaN.
```

```python
employees.dropna(how="all")  #Removes a row only when every value is NaN.
```

```python
employees.dropna(subset=["Salary"])
```

## fillna()

**Definition:**
Replaces missing values (`NaN`) with a specified value or method.

**Syntax:**
dataframe.fillna(value)

dataframe.fillna(value={"column": value})

dataframe.fillna(method="ffill")

dataframe.fillna(method="bfill")

```python
import pandas as pd
import numpy as np

employees = pd.DataFrame({
    "Employee": ["Amit", "Neha", "Rahul", "Priya", "Karan"],
    "Department": ["HR", "IT", np.nan, "IT", "Finance"],
    "Salary": [45000, np.nan, 55000, 80000, np.nan],
    "Experience": [2, 5, 3, np.nan, 4]
})

employees
```

```python
#Fill All Missing Values with a Single Value
employees.fillna(0)
```

```python
employees["Salary"].fillna(60000)   #Fill Only One Column
```

```python
employees.fillna({       #Fill Different Columns with Different Values
    "Department":"Unknown",
    "Salary":50000,
    "Experience":0
    
})
```

```python
#Forward Fill (ffill)
employees.ffill()
```

```python
#Backward Fill (bfill)
employees.bfill()
```

## replace()

**Definition:**
Replaces one or more specified values in a DataFrame or Series with new values.

**Syntax:**
dataframe.replace(old_value, new_value)

dataframe.replace({old1: new1, old2: new2})

```python
import pandas as pd

employees = pd.DataFrame({
    "Employee": ["Amit", "Neha", "Rahul", "Priya", "Karan"],
    "Department": ["HR", "IT", "NA", "IT", "Finance"],
    "Salary": [45000, 0, 55000, 80000, 0],
    "Experience": [2, 5, 3, 7, 4]
})

employees
```

```python
#Replace a single Value 
employees.replace(0,50000)
```

```python
import numpy as np
import pandas as pd
employees.replace("NA","Unknown")
```

```python
employees.replace({
    "NA":"Unknown",
    0:5000
})
```

```python
employees["Salary"].replace(0,50000)
```
