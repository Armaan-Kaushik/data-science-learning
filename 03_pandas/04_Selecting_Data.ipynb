```python
import pandas as pd

employees = {
    "Employee_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Priya", "Rahul", "Sneha", "Vikas", "Neha"],
    "Department": ["HR", "IT", "Finance", "IT", "Sales", "HR"],
    "Age": [25, 30, 28, 35, 40, 27],
    "Salary": [35000, 60000, 50000, 75000, 45000, 38000]
}

df = pd.DataFrame(employees)

df
```

# 📌 Selecting a Single Column

> **Definition:** Selecting a single column means retrieving one specific column from a DataFrame. In Pandas, this is done by passing the column name inside a single pair of square brackets (`[]`).

```python
#syntax:-  df["Column Name"]
df["Name"]
```

```python
df["Salary"]
```

```python
print(df["Salary"].mean())
print(df["Age"].max())
print(df["Name"].count())
```

# 📌 Selecting Multiple Columns

> **Definition:** Selecting multiple columns means retrieving two or more columns from a DataFrame. In Pandas, this is done by passing a **list of column names** inside **double square brackets (`[[]]`)**.

```python
# Syntax:-  df[["Column1", "Column2", "Column3"]]
print(df[["Age","Salary"]])
print(df[["Age","Salary"]].mean())
```

# 📌 Selecting Rows using Slicing

> **Definition:** Row slicing is the process of selecting a range of rows from a DataFrame using Python slicing syntax (`start:stop`). The start index is included, while the stop index is excluded.

```python
# Syntax:- df[start:stop]
print(df[1:4])
print(df[:3])
print(df[3:])
```

# 📌 loc[] – Label-Based Indexing

> **Definition:** `loc[]` is a Pandas indexer used to select rows and columns by their **labels (index names and column names)**. Unlike Python slicing, **the stop label is included** in `loc[]`.

```python
# Syntax:-  df.loc[row_selection, column_selection]
print(df.loc[2])
```

```python
print(df.loc[1:4])  # loc[] includes stop value whereas slicing don't include it.
```

```python
print(df.loc[:,["Name","Salary"]])
```

```python
print(df.loc[1:4, ["Name", "Salary"]])
```

# 📌 iloc[] – Position-Based Indexing

> **Definition:** `iloc[]` is a Pandas indexer used to select rows and columns based on their **integer positions**. It follows Python slicing rules, where the **start position is included** and the **stop position is excluded**.

```python
# Syntax:- df.iloc[row_selection, column_selection]
print(df.iloc[3])
```

```python
print(df.iloc[2])
```

```python
print(df.iloc[1:4])
```

```python
print(df.iloc[:,[1,4]])
```

```python
print(df.iloc[1:4, [1, 4]])
```

```python
df.iloc[1:4, 0:2]
```

# 📌 Boolean Indexing (Filtering Data)

> **Definition:** Boolean Indexing is the process of selecting rows from a DataFrame based on one or more conditions. It returns only the rows where the condition evaluates to **True**.

```python
# Syntax:-   df[condition]
print(df[df["Salary"]>50000])
print(df[df["Age"]>30])
print(df[df["Age"]<30])
```

```python
print(df[df["Department"] == "IT"])
```

```python
print(df[(df["Salary"] > 40000) & (df["Age"] < 35)])
```

```python
print(df[(df["Department"] == "HR") | (df["Department"] == "Sales")])
```

# 📌 Selecting Columns using Dot Notation (`df.ColumnName`)

> **Definition:** Dot notation is a shortcut for selecting a single column from a DataFrame by writing the column name after a dot (`.`). It can only be used when the column name is a valid Python identifier.

```python
#Syntax:- df.ColumnName
print(df.Name)
```

```python
print(df.Employee ID)
# this willl produce an error as it is not a valid python identifier so we should
#try to not use this method much 
```

# 📌 at[] – Accessing a Single Value

> **Definition:** `at[]` is a Pandas indexer used to access or update a **single value** in a DataFrame using a **row label** and a **column name**. It is faster than `loc[]` when working with only one cell.

```python
# Syntax:- df.at[row_label, column_name]
print(df.at[3,"Salary"]) #Gives salary at index 3rd 
```

```python
df.at[4, "Department"]
```

```python
# at[] also modifies a single value 
df.at[2,"Salary"]= 55000
print(df)
```

# 📌 iat[] – Accessing a Single Value by Position

> **Definition:** `iat[]` is a Pandas indexer used to access or update a **single value** in a DataFrame using the **integer position** of the row and column. It is optimized for fast access to one specific cell by position.

```python
# Syntax:- df.iat[row_position, column_position]
print(df.iat[3,4])
```

```python
print(df.iat[4, 2])
```

```python
df.iat[2, 4] = 50000
print(df)
```

## 📊 Comparison of `loc[]`, `iloc[]`, `at[]`, and `iat[]`

| Feature | `loc[]` | `iloc[]` | `at[]` | `iat[]` |
|---------|---------|----------|---------|----------|
| **Selection Based On** | Labels | Integer Positions | Labels | Integer Positions |
| **Row Selection** | ✅ Yes | ✅ Yes | ❌ Single Row Only | ❌ Single Row Only |
| **Column Selection** | ✅ Yes | ✅ Yes | ❌ Single Column Only | ❌ Single Column Only |
| **Multiple Rows** | ✅ Yes | ✅ Yes | ❌ No | ❌ No |
| **Multiple Columns** | ✅ Yes | ✅ Yes | ❌ No | ❌ No |
| **Single Value** | ✅ Yes | ✅ Yes | ✅ Optimized | ✅ Optimized |
| **Update Values** | ✅ Yes | ✅ Yes | ✅ Yes | ✅ Yes |
| **Faster for One Cell** | ❌ No | ❌ No | ✅ Yes | ✅ Yes |

---

## 📝 Syntax Comparison

| Indexer | Syntax | Example |
|---------|--------|---------|
| `loc[]` | `df.loc[row_label, column_name]` | `df.loc[2, "Salary"]` |
| `iloc[]` | `df.iloc[row_position, column_position]` | `df.iloc[2, 4]` |
| `at[]` | `df.at[row_label, column_name]` | `df.at[2, "Salary"]` |
| `iat[]` | `df.iat[row_position, column_position]` | `df.iat[2, 4]` |

---

## 🎯 Which One Should You Use?

| Situation | Recommended Indexer |
|-----------|---------------------|
| Select multiple rows/columns using **labels** | `loc[]` |
| Select multiple rows/columns using **positions** | `iloc[]` |
| Access or update **one value** using **labels** | `at[]` |
| Access or update **one value** using **positions** | `iat[]` |
