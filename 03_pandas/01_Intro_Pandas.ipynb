The simplest object in Pandas is called a Series.

A Series is a one-dimensional labeled array.

```python
import pandas as pd
marks=pd.Series([65,72,81,90])
print(marks)
```

Pandas automatically generated an index starting from 0.

Labels      Values

0 --------> 65
1 --------> 72
2 --------> 81
3 --------> 90

## Creating a Series with Custom Index

By default, Pandas creates numeric indexes. We can also assign meaningful labels to make our data easier to understand and access.

```python
sales=pd.Series(
    [500,650,720],
    index=["Jan","Feb","Mar"]
)
print(sales)
```

## DataFrame

A DataFrame is the most important data structure in Pandas. It stores data in rows and columns, just like an Excel spreadsheet or a SQL table.

```python
employees={
    "Name":["Rahul","Armaan","Amit"],
    "Department":["IT","HR","Finance"],
    "Salary":[65000,85000,72000]
}
df=pd.DataFrame(employees)
print(df)
```

## Understanding Series Attributes

After creating a Series, we often want to know information about it, such as its data type, size, index, or values. Pandas provides built-in attributes to access this information easily.

```python
import pandas as pd
marks=pd.Series([65,72,81,90])
print(marks)
```

```python
print(marks.dtype) # Returns the datatype of the Series
print(marks.size) # size returns the total number of elements in the Series.
print(marks.shape) # shape returns the dimensions of the Series.
print(marks.index) # This tells us how Pandas has indexed the data.
print(marks.values) # values returns the underlying NumPy array stored inside the Series.
print(marks.name) 
```

## Giving a Name to a Series

A Series can also have a name. This is useful when the Series represents a specific column, such as Salary, Marks, or Age.

```python
marks = pd.Series([65, 72, 81, 90], name="Marks")
print(marks)
```

## Different Ways to Create a Series

A Series can be created from different Python objects such as lists, tuples, NumPy arrays, dictionaries, and even a single scalar value. Knowing all these methods helps you work with data from different sources.

```python
#Creating a series from python list
import pandas as pd
marks=[85,90,78,88]
series_marks= pd.Series(marks)
print(series_marks)
```

```python
#Creating a series from tuple
price=(250,450,600,800)
series_price=pd.Series(price)
print(series_price)
```

```python
#Creating a series from numpy array
import numpy as np
import pandas as pd
scores=np.array([91,84,76,95])
score_series = pd.Series(scores)
print(score_series)
```

```python
#Creating a series from dictionary
sales={
    "Jan":500,
    "Feb":650,
    "Mar":720
}
sales_series = pd.Series(sales)
print(sales_series)
#Here dictionary keys becomes index
```

```python
#Creating a series from scalar Value
import pandas as pd
bonus= pd.Series(
    5000,
    index=["Emp1","Emp2","Emp3","Emp4"]
)
print(bonus)
```

```python
#Creating an Empty Series
empty_series=pd.Series(dtype="float64")
print(empty_series)
```

If you create an empty Series without specifying dtype, Pandas may show a warning in some versions. Therefore, always specify the data type for an empty Series.

## Series vs Python List vs NumPy Array

Python Lists, NumPy Arrays, and Pandas Series can all store one-dimensional data, but they are designed for different purposes. Understanding their differences helps you choose the right data structure for the right task.

```python
marks = pd.Series(
    [85, 90, 78],
    index=["Amit", "Rahul", "Priya"]
)

print(marks["Rahul"])
```

```python
import pandas as pd
marks = pd.Series([85, 90, 78])
print(marks + 10)
```

```python
#Pandas can store multiple datatype objects in a single array
employee=pd.Series([101,"Rahul",65000])
print(employee)
```

Because different types are mixed together, Pandas stores them as object.

```python
import pandas as pd
employees = {
    "Name": ["Rahul", "Priya", "Amit"],
    "Department": ["IT", "HR", "Finance"],
    "Salary": [65000, 58000, 72000]
}
df = pd.DataFrame(employees)
print(df)
```

```python
print(df.shape)
print(df.size)
print(df.columns)  #df.columns returns an Index object, not a list.
print(df.columns.tolist()) # to convert columns values into a Python list.
print(df.index)
print(df.dtypes)
print(df.values)
print(df.axes)  #The axes attribute returns both the row index and the column index of the DataFrame.
```

## Different Ways to Create a DataFrame

```python
# Dictionary of Lists
import pandas as pd
employees={
    "Name":["Rahul","Priya","Amit"],
    "Department":["IT","HR","Finance"],
    "Salary":["65000","58000","72000"]
}
df=pd.DataFrame(employees)
print(df)
```

```python
#List of Lists
employee_data=[
    ["Rahul","IT",65000],
    ["Priya","HR",58000],
    ["Amit","Finance",72000]
]
df= pd.DataFrame(
    employee_data,
    columns=["Name","Department","Salary"]
)
print(df)
```

```python
#List of Dictionaries
employee_data = [
    {"Name": "Rahul", "Department": "IT", "Salary": 65000},
    {"Name": "Priya", "Department": "HR", "Salary": 58000},
    {"Name": "Amit", "Department": "Finance", "Salary": 72000}
]
df = pd.DataFrame(employee_data)
print(df)
```

```python
#NumPy Array
import numpy as np
data=np.array([
    ["Rahul", "IT", 65000],
    ["Priya", "HR", 58000],
    ["Amit", "Finance", 72000]
])
df=pd.DataFrame(
    data,
    columns=["Name","Department","Salary"]
)
print(df)
```

```python
#Dictionary of Series
names = pd.Series(["Rahul", "Priya", "Amit"])
department = pd.Series(["IT", "HR", "Finance"])
salary = pd.Series([65000, 58000, 72000])
df=pd.DataFrame({
    "Name":names,
    "Department":department,
    "Salary":salary
})
print(df)
```

| Method | Used In |
|--------|---------|
| Dictionary of Lists | Most common way to create DataFrames from Python data |
| List of Lists | Data available as rows (Database records, CSV rows) |
| List of Dictionaries | JSON data, REST APIs, Web responses |
| NumPy Array | Machine Learning, Scientific Computing, Numerical Data |
| Dictionary of Series | Combining existing Pandas Series into a DataFrame |
