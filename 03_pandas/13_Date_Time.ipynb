# Datetime

Datetime is a special Pandas data type used to store and work with dates and times. Converting date columns to datetime allows you to filter, sort, calculate differences, and extract date components efficiently.

```python
import pandas as pd
df = pd.DataFrame({
    "Order_ID": [101, 102, 103, 104],
    "Order_Date": [
        "2025-01-15",
        "2025-02-10",
        "2025-03-05",
        "2025-04-20"
    ],
    "Sales": [1200, 2500, 1800, 3000]
})
df
```

```python
df.dtypes
```

```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"])  #Convert to Datetime
```

```python
df.dtypes
```

# Date Extraction

The `.dt` accessor is used to extract different components from a datetime column, such as the year, month, day, weekday, and more.

```python
import pandas as pd
df = pd.DataFrame({
    "Order_ID": [101, 102, 103, 104],
    "Order_Date": [
        "2025-01-15",
        "2025-02-10",
        "2025-03-05",
        "2025-04-20"
    ],
    "Sales": [1200, 2500, 1800, 3000]
})

df["Order_Date"] = pd.to_datetime(df["Order_Date"])
df
```

```python
df["Order_Date"].dt.year
```

```python
df["Year"] = df["Order_Date"].dt.year
df
```

```python
df["Order_Date"].dt.month
```

```python
df["Month"] = df["Order_Date"].dt.month
```

```python
df["Order_Date"].dt.day
```

```python
df["Day"] = df["Order_Date"].dt.day
```

```python
df["Order_Date"].dt.day_name()
```

```python
df["Weekday"] = df["Order_Date"].dt.day_name()
```

```python
df["Month_Name"] = df["Order_Date"].dt.month_name()
```

```python
df
```

# Date Filtering

Date filtering is used to select rows based on specific dates or date ranges. It helps answer business questions such as finding recent orders, monthly sales, or transactions within a given period.

```python
import pandas as pd
df = pd.DataFrame({
    "Order_ID": [101,102,103,104,105,106],
    "Order_Date":[
        "2025-01-15",
        "2025-02-10",
        "2025-03-05",
        "2025-04-20",
        "2025-05-12",
        "2025-06-18"
    ],
    "Sales":[1200,2500,1800,3000,2700,1500]
})
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
df
```

```python
df[df["Order_Date"] > "2025-03-01"]
```

```python
df[df["Order_Date"] < "2025-03-01"]
```

```python
df[df["Order_Date"] == "2025-04-20"]
```

```python
df[
    (df["Order_Date"] >= "2025-02-01") &
    (df["Order_Date"] <= "2025-05-01")
]
```

```python
df[df["Order_Date"].dt.month == 3]
```

```python
df[df["Order_Date"].dt.year == 2025]
```

```python
df[df["Order_Date"].dt.day_name() == "Monday"]
```

# Time Calculations

Time calculations are used to add, subtract, and compare dates. They help calculate durations such as delivery time, employee experience, payment due dates, and days between events.

```python
import pandas as pd
df = pd.DataFrame({
    "Order_ID": [101, 102, 103, 104],
    "Order_Date": [
        "2025-01-15",
        "2025-02-10",
        "2025-03-05",
        "2025-04-20"
    ],
    "Delivery_Date": [
        "2025-01-18",
        "2025-02-15",
        "2025-03-09",
        "2025-04-25"
    ]
})
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
df["Delivery_Date"] = pd.to_datetime(df["Delivery_Date"])
df
```

```python
df["Delivery_Date"] - df["Order_Date"]
```

```python
df["Delivery_Time"] = df["Delivery_Date"] - df["Order_Date"]
df
```

```python
df["Due_Date"] = df["Order_Date"] + pd.Timedelta(days=7)
df     #Add Days to a Date
```

```python
df["Reminder_Date"] = df["Order_Date"] - pd.Timedelta(days=2)
df   #Subtract Days from a date
```

# Real-World Date Analysis

Real-world date analysis involves using datetime operations to answer business questions such as monthly sales, delivery performance, weekday trends, and recent transactions.

```python
import pandas as pd
df = pd.DataFrame({
    "Order_ID": [101,102,103,104,105,106],
    "Order_Date": [
        "2025-01-15",
        "2025-02-10",
        "2025-03-05",
        "2025-03-22",
        "2025-04-18",
        "2025-05-12"
    ],
    "Delivery_Date": [
        "2025-01-18",
        "2025-02-14",
        "2025-03-08",
        "2025-03-28",
        "2025-04-22",
        "2025-05-18"
    ],
    "Sales": [1200,2500,1800,3200,2700,4000]
})

df["Order_Date"] = pd.to_datetime(df["Order_Date"])
df["Delivery_Date"] = pd.to_datetime(df["Delivery_Date"])
df
```

```python
#no of orders placed in each month?
df["Month"] = df["Order_Date"].dt.month_name()
df.groupby("Month")["Order_ID"].count()
```

```python
#no of orders placed in each month?
df.groupby("Month")["Sales"].sum()
```

```python
# average delivery time
df["Delivery_Days"] = (
    df["Delivery_Date"] -
    df["Order_Date"]
).dt.days
```

```python
df["Delivery_Days"].mean()
```

```python
df[df["Delivery_Days"] > 5]
```

```python
# Which weekday receives the most orders
df["Weekday"] = df["Order_Date"].dt.day_name()
df.groupby("Weekday")["Order_ID"].count()
```

```python
# all orders after March 1, 2025
df[df["Order_Date"] > "2025-03-01"]
```

```python
# Which month had the highest sales
df.groupby("Month")["Sales"].sum().sort_values(ascending=False)
```
