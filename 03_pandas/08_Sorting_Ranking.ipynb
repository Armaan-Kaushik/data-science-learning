```python
import pandas as pd
employees = pd.DataFrame({
    "Emp_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Neha", "Rahul", "Sneha", "Karan", "Priya"],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "Finance"],
    "Age": [25, 30, 28, 35, 26, 32],
    "Salary": [45000, 70000, 65000, 85000, 48000, 72000]
})
employees
```

# Definition: sort_values()

`sort_values()` is a Pandas method used to arrange the rows of a DataFrame based on the values of one or more columns in either ascending or descending order.

```python
employees.sort_values(by="Salary")
```

```python
employees.sort_values(by="Salary",ascending=False)
```

```python
employees.sort_values(
    by=["Department", "Salary"],
    ascending=[True, False]
)
```

```python
employees.sort_values(by="Salary",ascending=False,ignore_index=True)
```

```python
employee_salary = pd.DataFrame({
    "Name": ["Amit", "Neha", "Rahul", "Sneha"],
    "Salary": [45000, None, 65000, 85000]
})

employee_salary
```

```python
employee_salary.sort_values(by="Salary",na_position="first")
```

# Definition: sort_index()

`sort_index()` is a Pandas method used to arrange a DataFrame based on its row index or column index instead of the values stored inside the DataFrame.

```python
employees = pd.DataFrame({
    "Name": ["Amit", "Neha", "Rahul", "Sneha"],
    "Salary": [45000, 70000, 65000, 85000]
},
index=[104, 101, 103, 102])

employees
```

```python
employees.sort_index()
```

```python
employees.sort_index(ascending=False)
```

```python
employee_data = pd.DataFrame({
    "Salary": [45000, 70000],
    "Age": [25, 30],
    "Department": ["HR", "IT"],
    "Name": ["Amit", "Neha"]
})
employee_data
```

```python
employee_data.sort_index(axis=1)
```

# Definition: rank()

`rank()` is a Pandas method used to assign a rank (position) to each value in a Series or DataFrame based on its magnitude.

```python
import pandas as pd
employees = pd.DataFrame({
    "Name": ["Amit", "Neha", "Rahul", "Sneha", "Karan"],
    "Salary": [45000, 70000, 65000, 85000, 48000]
})
employees
```

```python
employees["Salary"].rank(ascending=False)
```

```python
employees["Rank"]=employees["Salary"].rank(ascending=False)
employees
```

```python
employees = pd.DataFrame({
    "Name": ["Amit", "Neha", "Rahul", "Sneha"],
    "Salary": [70000, 70000, 50000, 90000]
})
employees
```

```python
employees["Salary"].rank(ascending=False)
```

# Different Ranking Methods

```python
employees["Salary"].rank(
    ascending=False,
    method="average"     # Default method
)
```

```python
employees["Salary"].rank(
    ascending=False,
    method="min"   # All tied values receive the smallest possible rank
)
```

```python
employees["Salary"].rank(
    ascending=False,
    method="max"    # All tied values receive the largest possible rank.
)
```

```python
employees["Salary"].rank(
    ascending=False,
    method="first"  #if values are equal,rank them according to their order of appearance.
)
```

```python
employees["Salary"].rank(
    ascending=False,
    method="dense"
)  
```
