# pivot()

`pivot()` is a Pandas method used to reshape a DataFrame by converting unique values from one column into new columns. It does **not** perform any aggregation.It does not woro with duplicate combinations.

df.pivot(

    index=None,
    
    columns=None,
    
    values=None
)

```python
import pandas as pd
sales = pd.DataFrame({
    "Region": ["North", "North", "South", "South"],
    "Product": ["Laptop", "Phone", "Laptop", "Phone"],
    "Sales": [50000, 30000, 40000, 35000]
})
sales
```

```python
sales.pivot(
    index="Region",
    columns="Product",
    values="Sales"
)
```

# pivot_table()

`pivot_table()` is a Pandas method used to create Excel-like pivot tables by reshaping data and applying an aggregation function to duplicate values.

df.pivot_table(
    
    index=None,
    
    columns=None,
    
    values=None,
    
    aggfunc="mean"
)

```python
sales = pd.DataFrame({
    "Region": ["North","North","North","South","South"],
    "Product": ["Laptop","Laptop","Phone","Laptop","Phone"],
    "Sales": [50000,55000,30000,40000,35000]
})
sales
```

```python
sales.pivot_table(
    index="Region",
    columns="Product",
    values="Sales",
    aggfunc="sum"  # Default agg function is mean
)
```

```python
sales.pivot_table(
    index="Region",
    columns="Product",
    values="Sales",
    aggfunc="count" 
)
```

```python
sales.pivot_table(
    index="Region",
    columns="Product",
    values="Sales",
    aggfunc="max"
)
```

```python
sales.pivot_table(
    index="Region",
    columns="Product",
    values="Sales",
    aggfunc="min"
)
```

```python
sales.pivot_table(
    index="Region",
    columns="Product",
    values="Sales",
    aggfunc=["sum","mean","max","count"]
)
```

# Multiple Indexes in pivot_table()

`pivot_table()` allows you to use multiple columns as the row index, creating a hierarchical (MultiIndex) table for more detailed data analysis.


```python
import pandas as pd

sales = pd.DataFrame({
    "Region": [
        "North", "North", "North",
        "South", "South", "South",
        "East", "East"
    ],
    "Salesperson": [
        "Amit", "Amit", "Neha",
        "Rahul", "Rahul", "Sneha",
        "Priya", "Priya"
    ],
    "Product": [
        "Laptop", "Phone", "Laptop",
        "Laptop", "Phone", "Laptop",
        "Laptop", "Phone"
    ],
    "Sales": [
        50000, 30000, 55000,
        40000, 35000, 45000,
        60000, 25000
    ]
})

sales
```

```python
sales.pivot_table(
    index=["Region","Salesperson"],
    columns="Product",
    values="Sales",
    aggfunc="sum"
)
```

```python
sales.pivot_table(
    index="Region",
    columns=["Product", "Salesperson"],
    values="Sales",
    aggfunc="sum"
)
```

# fill_value

The `fill_value` parameter replaces missing values (`NaN`) in a pivot table with a specified value.

```python
sales.pivot_table(
    index=["Region","Salesperson"],
    columns="Product",
    values="Sales",
    aggfunc="sum"
)
```

```python
sales.pivot_table(
    index=["Region","Salesperson"],
    columns="Product",
    values="Sales",
    aggfunc="sum",
    fill_value=0
)
```

```python
sales.pivot_table(
    index="Region",
    columns="Product",
    values="Sales",
    aggfunc="sum",
    fill_value="-"
)
```

# margins=True

The `margins=True` parameter adds grand totals for rows and columns in a pivot table.

```python
sales.pivot_table(
    index="Region",
    columns="Product",
    values="Sales",
    aggfunc="sum",
    margins=True
)
```

```python
sales.pivot_table(
    index="Region",
    columns="Product",
    values="Sales",
    aggfunc="sum",
    margins=True,
    margins_name="Grand Total"   #To rename the total column
    )
```

```python
sales.pivot_table(
    index=["Region","Salesperson"],
    columns="Product",
    values="Sales",
    aggfunc="sum",
    margins=True
)
```

# crosstab()

`crosstab()` is a Pandas function used to create a frequency table that shows how often combinations of categorical variables occur.

pd.crosstab(
    index,
    columns
)

```python
employees = pd.DataFrame({
    "Department": [
        "HR", "HR", "IT", "IT",
        "Finance", "Finance", "IT", "HR"
    ],
    "Gender": [
        "Male", "Female", "Male", "Female",
        "Male", "Female", "Male", "Female"
    ]
})

employees
```

```python
pd.crosstab(employees["Department"],employees["Gender"])
```

```python
pd.crosstab(employees["Department"],
            employees["Gender"],
           margins="True",
           )
```

```python
pd.crosstab(employees["Department"],
            employees["Gender"],
            normalize="index"
)
           
```

```python
pd.crosstab(
    employees["Department"],
    employees["Gender"],
    normalize="columns"
)
```

```python
pd.crosstab(
    employees["Department"],
    employees["Gender"],
    normalize=True
)
```
