## Dataset Description

In this notebook, we'll work with a **sample Employee Performance Dataset** to simulate a real-world Data Science workflow using **NumPy**.

Each row represents one employee, and each column contains a specific attribute about that employee.

| Column | Description |
|--------|-------------|
| Employee_ID | Unique ID assigned to each employee |
| Age | Employee's age (in years) |
| Department | Department code where the employee works |
| Experience | Total years of work experience |
| Salary | Monthly salary (₹) |
| Sales | Monthly sales generated (₹) |
| Rating | Performance rating (1–5) |

### Department Codes

| Code | Department |
|------|------------|
| 1 | Sales |
| 2 | Marketing |
| 3 | Human Resources (HR) |
| 4 | Information Technology (IT) |

```python
import numpy as np
```

```python
import numpy as np

employee_data = np.array([
    [101, 25, 1, 2, 35000, 180000, 4],
    [102, 30, 2, 5, 50000, 220000, 5],
    [103, 28, 1, 4, 42000, 210000, 4],
    [104, 35, 4, 10, 75000, 300000, 5],
    [105, 26, 3, 3, 38000, 150000, 3],
    [106, 40, 4, 15, 90000, 420000, 5],
    [107, 29, 2, 6, 52000, 240000, 4],
    [108, 31, 1, 7, 60000, 260000, 5],
    [109, 27, 3, 4, 41000, 170000, 3],
    [110, 38, 4, 12, 82000, 390000, 5]
])

np.savetxt(
    "employee_data.csv",
    employee_data,
    delimiter=",",
    fmt="%d"
)

print("Dataset saved successfully as 'employee_data.csv'")
```

```python
data = np.loadtxt("employee_data.csv", delimiter=",", dtype=int)

print(data)
```

```python
print("Shape:", data.shape)
print("Dimensions:", data.ndim)
print("Data Type:", data.dtype)
print("Total Values:", data.size)
```

```python
employee_id = data[:, 0]
age = data[:, 1]
department = data[:, 2]
experience = data[:, 3]
salary = data[:, 4]
sales = data[:, 5]
rating = data[:, 6]
```

```python
print("Employee IDs:\n", employee_id)

print("\nSalaries:\n", salary)

print("\nSales:\n", sales)
```

```python
print(data[:, [1, 4, 5]])
```

```python
high_salary = data[salary > 50000]

print(high_salary)
```

```python
experienced_emp = data[experience > 5]

print(experienced_emp)
```

```python
top_rated = data[rating == 5]

print(top_rated)
```

```python
it_department = data[department == 4]

print(it_department)
```

```python
high_performers = data[(salary > 50000) & (rating == 5)]

print(high_performers)
```

```python
print("Average Salary :", np.mean(salary))
print("Median Salary  :", np.median(salary))
print("Highest Salary :", np.max(salary))
print("Lowest Salary  :", np.min(salary))
print("Salary Std Dev :", np.std(salary))
```

```python
print("Average Sales :", np.mean(sales))
print("Median Sales  :", np.median(sales))
print("Highest Sales :", np.max(sales))
print("Lowest Sales  :", np.min(sales))
print("Sales Std Dev :", np.std(sales))
```

```python
print("Average Experience :", np.mean(experience))
print("Maximum Experience :", np.max(experience))
print("Minimum Experience :", np.min(experience))
```

```python
salary_sorted = data[np.argsort(salary)]

print(salary_sorted)
```

```python
annual_salary = salary * 12

print(annual_salary)
```

```python
annual_salary = annual_salary.reshape(-1,1 )  
#Here, -1 means:" NumPy, you figure out how many rows are needed."

print(annual_salary)
print(np.shape(annual_salary))
```

```python
print("Average Age:", np.mean(age))
```

```python
highest_salary_employee = data[np.argmax(salary)]

print(highest_salary_employee)
```

```python
below_average_salary = data[salary < np.mean(salary)]

print(below_average_salary)
```

```python
updated_data = np.hstack((data, annual_salary))

print(updated_data)
```

```python
np.savetxt(
    "employee_data_processed.csv",
    updated_data,
    delimiter=",",
    fmt="%d"
)

print("Processed dataset saved successfully!")
```
