# 🔷 sort()

sort() returns a sorted copy of a NumPy array.

By default, it sorts the array in ascending order without changing the original array.

```python
import numpy as np

arr = np.array([8, 3, 1, 6, 4])
result=np.sort(arr)
print("Original Array:" ,arr)
print("Sorted Array:", result)
```

```python
import numpy as np

arr = np.array([[7, 2, 5],
                [9, 1, 3]])
result = np.sort(arr)
print(result)
```

```python
import numpy as np

arr = np.array([[7, 2, 5],
                [9, 1, 3]])
result = np.sort(arr, axis=0)
print(result)
```

```python
import numpy as np

arr = np.array([8, 3, 1, 6, 4])
print(np.sort(arr))
print(np.sort(arr)[::-1])
```

# 🔷 argsort()

argsort() returns the indices that would sort an array.

It tells you the positions of elements in sorted order instead of returning the sorted values.

```python
import numpy as np

arr = np.array([50, 20, 40, 10, 30])
result =np.argsort(arr)
print(result)
```

```python
import numpy as np

arr = np.array([50, 20, 40, 10, 30])
sorted_arr = arr[np.argsort(arr)]
print(sorted_arr)
```

```python
import numpy as np

names = np.array(["Rahul", "Amit", "Neha", "Priya"])
marks = np.array([75, 92, 68, 85])
index = np.argsort(marks)
print(index)
print(names[index])
print(marks[index])
```

# 🔷 searchsorted()

searchsorted() finds the index where a value should be inserted in a sorted array to maintain the sorted order.

The array must be sorted before using `searchsorted()`.

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])
result=np.searchsorted(arr,35)
print(result)
```

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])
result = np.searchsorted(arr, 30)
print(result)
```

```python
import numpy as np

arr = np.array([10, 20, 20, 20, 30, 40])

left = np.searchsorted(arr, 20, side="left")
right = np.searchsorted(arr, 20, side="right")

print("Left :", left)
print("Right:", right)
```

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])
values = [15, 35, 55]
result = np.searchsorted(arr, values)
print(result)
```

# 🔷 unique()

unique() returns the unique (distinct) elements from a NumPy array.

The returned values are sorted in ascending order.

```python
import numpy as np

arr = np.array([10, 20, 10, 30, 20, 40])
result= np.unique(arr)
print(result)
```

```python
import numpy as np

arr = np.array([5, 2, 8, 2, 5, 1])
result = np.unique(arr)
print(result)
```

```python
import numpy as np

arr = np.array(["Apple", "Banana", "Apple", "Mango", "Banana"])
result = np.unique(arr)
print(result)
```

```python
import numpy as np

arr = np.array([10, 20, 10, 30, 20, 40])
unique_values,counts=np.unique(arr,return_counts=True)
print("Unique Values:",unique_values)
print("Counts:",counts)
```

# 🔷 isin()

isin() checks whether each element of an array exists in another array.

It returns a Boolean array containing `True` or `False`.

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])
result = np.isin(arr, [20, 40])
print(result)
```

```python
import numpy as np

students = np.array(["Rahul", "Amit", "Neha", "Priya"])
present = np.array(["Amit", "Priya"])
result = np.isin(students, present)
print(result)
```

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])
result = arr[np.isin(arr, [20, 40])]
print(result)
```

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])
result = arr[~np.isin(arr, [20, 40])]
print(result)
```

| Function | Purpose | Returns |
