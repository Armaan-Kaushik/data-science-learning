# 🔷 Broadcasting

Broadcasting allows NumPy to perform element-wise operations on arrays of different shapes by automatically expanding the smaller array.

The arrays must satisfy NumPy's broadcasting rules to be compatible.

```python
import numpy as np

a = np.array([10, 20, 30])
print(a + 5)
```

```python
import numpy as np

a = np.array([[1],
              [2],
              [3]])

b = np.array([10, 20, 30])

print(a)
print(a.shape)

print("\n",b)
print(b.shape)

print(a + b)
```

```python
import numpy as np

a = np.array([[1, 2, 3]])

b = np.array([[10],
              [20]])

print(a.shape)
print(b.shape)

print(a + b)
```

```python
import numpy as np

a = np.array([[1],
              [2],
              [3]])

b = np.array([10, 20, 30])

print(a.shape)
print(b.shape)

print(a + b)
```

# 🔷 Broadcasting Rules

NumPy follows these three rules to determine whether two arrays can be broadcast together.

### Rule 1
If the arrays have different numbers of dimensions, NumPy adds dimensions of size **1** to the **left** of the smaller-dimensional array.

### Rule 2
Starting from the **rightmost dimension**, NumPy compares the dimensions. Broadcasting is possible only if:
- The dimensions are equal, or
- One of the dimensions is **1**.

Otherwise, broadcasting fails.

### Rule 3
If a dimension is **1**, NumPy automatically expands (broadcasts) it to match the corresponding dimension of the other array.

**Note:** Broadcasting is virtual. NumPy does **not** create a physical copy of the data.

```python
import numpy as np

a = np.array([[1, 2, 3],
              [4, 5, 6]])

b = np.array([[10, 20],
              [30, 40]])

print("a shape:", a.shape)
print("b shape:", b.shape)

print(a + b)
```

```python
import numpy as np

marks = np.array([65, 72, 81, 90])
bonus = 5
final_marks = marks + bonus
print(final_marks)
```

```python
import numpy as np

prices = np.array([100, 250, 400, 150])
discount = 0.10
final_prices = prices - (prices * discount)
print(final_prices)
```

# 🔷 Common Errors

Broadcasting fails when two arrays have incompatible shapes.

The most common error is `ValueError: operands could not be broadcast together`.

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([10, 20])
print(a + b)
```
