# NumPy - Array Creation
## Creating Arrays in NumPy

## 1. np.array()

The `np.array()` function is used to convert Python sequences such as lists or tuples into NumPy arrays.

Syntax:

np.array(object)

```python
import numpy as np
numbers=[10,20,30,40,50]
arr=np.array(numbers)
print(arr)
```

```python
type(arr)
```

```python
numbers = (1,2,3,4)
arr = np.array(numbers)
print(arr)
```

```python
import numpy as np
a=np.array([1,2,3])
b=np.array((4,5,6))
c=np.array([[7,8],[9,10]])
print(a)
print(b)
print(c)
print(type(a))
print(type(b))
print(type(c))
```

## Key Takeaways

- `np.array()` converts Python sequences into NumPy arrays.
- It can convert lists, tuples, and nested lists.
- The output object type is `numpy.ndarray`.
- NumPy displays arrays differently from Python lists.

## 2. np.zeros()

`np.zeros()` creates an array filled with zeros.

Syntax:

np.zeros(shape)

```python
import numpy as np
arr=np.zeros(5)
print(arr)
arr.dtype
```

```python
arr=np.zeros(5,dtype=int)
print(arr)
```

```python
arr=np.zeros((2,3))
print(arr)
```

```python
a=np.zeros(3)
print(a)
```

```python
b=np.zeros((3,2))
print(b)
```

```python
c=np.zeros((2,2),dtype=int)
print(c)
```

## 3. np.ones()

The `np.ones()` function creates a new NumPy array filled with ones.

Syntax:

np.ones(shape)

```python
import numpy as np
arr = np.ones(5)
print(arr)
arr.dtype
```

```python
arr = np.ones(5,dtype=int)
print(arr)
arr.dtype
```

```python
arr=np.ones((3,4))
print(arr)
```

```python
arr=np.ones((2,3,4))
print(arr)
```

## 4. np.empty()

The `np.empty()` function creates an array without initializing its values.

It allocates memory for the array but does not fill it with any specific values.

```python
import numpy as np
arr=np.empty(5)
print(arr)
```

```python
arr = np.empty((2,3))

print(arr)
```

```python
a = np.empty(4)
print(a)
```

```python
b = np.empty((2,2))
print(b)
```

```python
c = np.empty(3)
c[:] = [100, 200, 300]
print(c)
```

```python
arr = np.array([10,20,30,40,50])
arr[:] = 99
print(arr)
```

```python
arr = np.array([10,20,30,40,50])
arr[1:4] = [100,200,300]
print(arr)
```

## 5. np.arange()

`np.arange()` creates a NumPy array containing evenly spaced values within a given range.

Syntax:

np.arange(start, stop, step)

```python
import numpy as np
a=np.arange(5)
print(a)
```

```python
b=np.arange(2,8)
print(b)
```

```python
arr=np.arange(2,11,2)
print(arr)
```

```python
arr=np.arange(10,0,-2)
print(arr)
```

## 6. np.linspace()

The `np.linspace()` function creates evenly spaced numbers between a start and stop value.

Unlike `np.arange()`, you specify the **number of values**, not the step size.

Syntax:

np.linspace(start, stop, num)

```python
import numpy as np
arr=np.linspace(1,5,5)
print(arr)
```

```python
arr=np.linspace(0,10,6)
print(arr)
```

```python
arr=np.linspace(1,10,5)
print(arr)
```

```python
a = np.linspace(1,5,5)
print(a)
b = np.linspace(0,20,5)
print(b)
c = np.linspace(100,200,6)
print(c)
```

```python
arr=np.linspace(0,1,50)
print(arr)
```

## 7. np.eye()

The `np.eye()` function creates an identity matrix.

An identity matrix is a square matrix in which:
- All diagonal elements are 1.
- All other elements are 0.

Syntax:

np.eye(N)

np.eye(N,M) can also create non-square matrices.

```python
import numpy as np
arr=np.eye(3)
print(arr)
```

```python
arr=np.eye(4)
print(arr)
```

```python
arr=np.eye(3,4)
print(arr)
```

## 8. np.identity()

The `np.identity()` function also creates an identity matrix.

Syntax:

np.identity(n)

It only create a square identity matrix

```python
arr=np.identity(3)
print(arr)
```

## 9. np.full()

The `np.full()` function creates an array filled with a specified value.

Syntax:

np.full(shape, fill_value)

```python
arr=np.full(5,7)
print(arr)
```

```python
arr=np.full((2,3),100)
print(arr)
```

```python
arr=np.full((3,3),5)
print(arr)
```
