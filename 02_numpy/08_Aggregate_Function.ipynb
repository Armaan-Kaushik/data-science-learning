# 🔷 sum()

`sum()` is used to calculate the total of array elements.

It can calculate the sum of the entire array or along a specific axis.

```python
import numpy as np

arr = np.array([10, 20, 30, 40])
print(np.sum(arr))
```

```python
import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])
print(np.sum(arr))
```

```python
import numpy as np

sales = np.array([100, 150, 200, 250])
total_sales = np.sum(sales)
print(total_sales)
```

# 🔷 mean()

`mean()` is used to calculate the average (arithmetic mean) of array elements.

It returns the sum of all elements divided by the total number of elements.

```python
import numpy as np

arr = np.array([10, 20, 30, 40])
print(np.mean(arr))
```

```python
import numpy as np

arr = np.array([[2, 4],
                [6, 8]])
print(arr.shape)
print(np.mean(arr))
```

```python
import numpy as np

marks = np.array([78, 85, 92, 95])
average_marks = np.mean(marks)
print(average_marks)
```

# 🔷 median()

`median()` returns the middle value of a sorted array.

If the array has an even number of elements, it returns the average of the two middle values.

```python
import numpy as np

arr=np.array([10,20,30,40,50]) 
print(np.median(arr))
```

```python
import numpy as np

arr = np.array([10, 20, 30, 40])
print(np.median(arr))
```

```python
import numpy as np

arr = np.array([40, 10, 50, 20, 30])
print(np.median(arr))
```

# 🔷 std()

`std()` calculates the standard deviation of array elements.

It measures how much the values are spread out from the mean.

```python
import numpy as np

arr = np.array([10, 10, 10, 10])
print(np.std(arr))
```

```python
import numpy as np

arr = np.array([2, 4, 6, 8])
print(np.std(arr))
```

```python
import numpy as np

marks = np.array([78, 80, 82, 79, 81])
print(np.round(np.std(marks), 3))
```

# 🔷 var()

`var()` calculates the variance of array elements.

Variance measures how far the values are spread from the mean. It is the square of the standard deviation.

```python
import numpy as np

arr = np.array([10, 10, 10, 10])
print(np.var(arr))
```

```python
import numpy as np

marks = np.array([78, 80, 82, 79, 81])
print(np.round(np.var(marks), 2))
```

# 🔷 min()

If no axis is specified, it finds the minimum from the entire array.

```python
import numpy as np

arr = np.array([12, 5, 18, 9])
print(np.min(arr))
```

```python
import numpy as np

arr = np.array([25, -10, 40, 15])
print(np.min(arr))
```

```python
import numpy as np

temperatures = np.array([32, 28, 35, 30, 27])
lowest = np.min(temperatures)
print(lowest)
```

# 🔷 argmin()

`argmin()` returns the index of the smallest element in a NumPy array.

It returns the position, not the value.

```python
import numpy as np

arr = np.array([12, 5, 18, 9])
print(np.argmin(arr))
```

```python
import numpy as np

arr = np.array([25, -10, 40, 15])
print(np.argmin(arr))
```

```python
import numpy as np

marks = np.array([78, 92, 65, 88, 95])
lowest_index = np.argmin(marks)
print(lowest_index)
print(marks[lowest_index])
```

# 🔷 Axis Operations

The `axis` parameter tells NumPy along which direction to perform an operation.

`axis=0` works column-wise, while `axis=1` works row-wise.

```python
import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])
print(arr)
print(np.sum(arr,axis=0))
print(np.sum(arr,axis=1))
```

```python
import numpy as np

arr = np.array([[10, 20, 30],
                [40, 50, 60]])

print(np.mean(arr, axis=0))
print(np.mean(arr, axis=1))
print(np.mean(arr))
```

The axis parameter is supported by almost all NumPy aggregation functions:

np.sum()

np.mean()

np.median()

np.std()

np.var()

np.min()

np.max()

np.argmin()

np.argmax()

# 🔷 cumsum()

`cumsum()` returns the cumulative (running) sum of array elements.

Each element is the sum of itself and all previous elements.

```python
import numpy as np

arr = np.array([1, 2, 3, 4])
print(np.cumsum(arr))
```

```python
import numpy as np

sales = np.array([100, 200, 150, 250])
print(np.cumsum(sales))
```

```python
import numpy as np

arr = np.array([[1, 2],
                [3, 4]])
print(np.cumsum(arr))
print(np.cumsum(arr,axis=0))
print(np.cumsum(arr,axis=1))
```

# 🔷 cumprod()

`cumprod()` returns the cumulative (running) product of array elements.

Each element is multiplied with all previous elements.

```python
import numpy as np

arr = np.array([1, 2, 3, 4])
print(np.cumprod(arr))
```

```python
import numpy as np

arr = np.array([2, 3, 4])
print(np.cumprod(arr))
```

```python
import numpy as np

arr = np.array([[1, 2],
                [3, 4]])
print(np.cumprod(arr))
print(np.cumprod(arr,axis=0))
print(np.cumprod(arr,axis=1))
```
