# 📖 What is `reshape()`?

The **`reshape()`** function is used to change the **shape (dimensions)** of an existing NumPy array **without changing its data**.

It rearranges the elements into a new shape while preserving all the values.

Think of it as reorganizing books on a shelf. The books remain the same; only their arrangement changes.

Similarly, `reshape()` only changes how the data is organized, not the data itself.

```python
import numpy as np

arr = np.arange(1, 13)

print("Original Array:")
print(arr)

reshaped_arr = arr.reshape(2, 2, 3)

print("\nReshaped Array:")
print(reshaped_arr)

print("\nShape:", reshaped_arr.shape)
```

```python
import numpy as np

arr = np.arange(1, 13)

print("Original Array:")
print(arr)

arr = arr.reshape(2, 2, 3)

print("\nReshaped Array:")
print(arr)

print("\nShape:",arr.shape)
```

```python
import numpy as np
arr = np.arange(1, 13)
arr.reshape(3, -1) #Automatic dimension using -1
```

```python
import numpy as np
arr = np.arange(1, 13)
arr.reshape(-1, 3)
```

```python
import numpy as np
arr = np.arange(1, 7)
arr.reshape(4, 2) #Error
```

```python
import numpy as np
arr=np.arange(1,17)
new_arr=arr.reshape(4,4)
print(new_arr)
```

# 🔷 flatten()

`flatten()` is used to convert a multi-dimensional NumPy array into a **1D array**.

It returns a **new copy** of the array, so changes made to the flattened array do **not** affect the original array.

```python
import numpy as np

arr = np.array([[10, 20, 30],
                [40, 50, 60]])

print("Original Array:")
print(arr)
flatten_arr=arr.flatten()
print(flatten_arr)
```

```python
import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])

flat = arr.flatten()
flat[0] = 100
print("Flattened Array:")
print(flat)
print("\nOriginal Array:")
print(arr)
```

```python
import numpy as np
arr=np.arange(1,13).reshape(2,2,3)
print(arr)
flat=arr.flatten()
print(flat)
```

# 🔷 ravel()

`ravel()` is used to convert a multi-dimensional NumPy array into a **1D array**.

Unlike `flatten()`, `ravel()` returns a **view** of the original array whenever possible, making it faster and more memory efficient.

```python
import numpy as np

arr = np.array([[10, 20, 30],
                [40, 50, 60]])

print("Original Array:")
print(arr)

ravel_arr = arr.ravel()

print("\nRaveled Array:")
print(ravel_arr)

print("\nShape:", ravel_arr.shape)
```

```python
import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])
ravel=arr.ravel()
ravel[0]=100
print("Raveled Array:")
print(ravel)
print("\nOriginal Array:")
print(arr)
```

```python
import numpy as np

arr = np.arange(1, 13).reshape(2, 2, 3)

print("Original Shape:", arr.shape)

ravel_arr = arr.ravel()

print("\nRaveled Array:")
print(ravel_arr)

print("\nNew Shape:", ravel_arr.shape)
```

```python
import numpy as np
arr = np.array([[1, 2],
                [3, 4]])

flat = arr.flatten()
rav = arr.ravel()

flat[0] = 100
rav[1] = 200

print("Original Array:")
print(arr)

print("\nFlatten:")
print(flat)

print("\nRavel:")
print(rav)
```

# 🔷 transpose()

`transpose()` is used to swap the axes (dimensions) of a NumPy array.

For a 2D array, it converts **rows into columns** and **columns into rows**.

```python
import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print("Original Array:")
print(arr)
print("\nOriginal Shape:", arr.shape)
transposed_arr=arr.transpose()
print("Transposed array")
print(transposed_arr)
print("\nTransposedd array shape:",transposed_arr.shape)
```

```python
import numpy as np

arr = np.array([[10, 20],
                [30, 40]])

print(arr.T) # Shortcut for transpose
```

# 🔷 resize()

`resize()` is used to change the shape **and size** of a NumPy array.

Unlike `reshape()`, `resize()` can **add or remove elements**. If new elements are needed, NumPy fills them by repeating the existing data.

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6])

print("Original Array:")
print(arr)

arr.resize((2, 3))

print("\nResized Array:")
print(arr)

print("\nShape:", arr.shape)
```

```python
import numpy as np

arr = np.array([1, 2, 3, 4])
arr.resize((3, 3))
print(arr)
```

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50, 60])
arr.resize((2, 2))
print(arr)
```

# 🔷 squeeze()

`squeeze()` is used to remove axes (dimensions) of size **1** from a NumPy array.

It is useful when working with arrays that contain unnecessary dimensions after reshaping or loading data.

```python
import numpy as np
arr = np.array([[[10, 20, 30]]])
print("Original Shape:", arr.shape)
squeezed_arr=np.squeeze(arr)
print("\nSqueezed array:",squeezed_arr)
print("\nNew Shape:",squeezed_arr.shape)
```

```python
import numpy as np

arr = np.array([[[1, 2, 3]]])

print("Original Shape:", arr.shape)

new_arr = np.squeeze(arr, axis=0)

print("\nShape:", new_arr.shape)

print(new_arr)
```

```python
import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])

np.squeeze(arr, axis=1) #Gives Error as axes 1 has more than 1 elements
```

# 🔷 expand_dims()

`expand_dims()` is used to add a new axis (dimension) to an existing NumPy array.

It is the opposite of `squeeze()` and is useful when preparing data for machine learning and deep learning models.

```python
import numpy as np

arr = np.array([10, 20, 30])
print("Original Shape:", arr.shape)
new_arr = np.expand_dims(arr, axis=0)

print("\nNew Array:")
print(new_arr)

print("\nNew Shape:", new_arr.shape)
```

```python
import numpy as np

arr = np.array([10, 20, 30])

new_arr = np.expand_dims(arr, axis=1)

print(new_arr)

print(new_arr.shape)
```

```python
import numpy as np

arr = np.array([[1, 2],
                [3, 4]])

print("Original Shape:", arr.shape)

new_arr = np.expand_dims(arr, axis=2)

print("\nNew Shape:", new_arr.shape)
```

```python
arr = np.array([10,20,30])

new_arr = arr[np.newaxis, np.newaxis, :]

print(new_arr.shape)
```
