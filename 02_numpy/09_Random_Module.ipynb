# 🔷 Random module()
# 🔷 rand()

`rand()` generates random numbers between **0 (inclusive)** and **1 (exclusive)**.

It returns floating-point numbers and is useful for creating random datasets.

```python
import numpy as np 
print(np.random.rand())
```

```python
import numpy as np
print(np.random.rand(5))
```

```python
import numpy as np
print(np.random.rand(2,3))
```

# 🔷 randint()

`randint()` generates random integers within a given range.

The starting value is included, but the ending value is excluded.

```python
import numpy as np
print(np.random.randint(1,10))
```

```python
import numpy as np
print(np.random.randint(1,10,size=5))
```

```python
import numpy as np
print(np.random.randint(1,10,size=(2,5)))
```

# 🔷 randn()

`randn()` generates random numbers from the **standard normal (Gaussian) distribution**.

The values are centered around **0** and can be positive or negative.

```python
import numpy as np 
print(np.random.randn())
```

```python
import numpy as np
print(np.random.randn(5))
```

```python
import numpy as np
print(np.random.randn(2,4))
```

# 🔷 choice()

`choice()` randomly selects one or more elements from an array or list.

It can select with or without replacement.

```python
import numpy as np
arr=np.array([10, 20, 30, 40, 50])
print(np.random.choice(arr))
```

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])
print(np.random.choice(arr,size=3))
```

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])
print(np.random.choice(arr,size=3,replace=False))

#replace=False means without replacement.
#Once an element is selected, it cannot be selected again.
#Therefore, all selected values are unique.
```

```python
import numpy as np

arr = np.array([1, 2, 3])

print(np.random.choice(arr, size=5, replace=False))
```

```python
import numpy as np

arr = np.array([1, 2, 3])

print(np.random.choice(arr, size=5))
```

# 🔷 shuffle()

`shuffle()` randomly rearranges the elements of an array.

It modifies the original array (in-place) and does not create a new one.

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])
np.random.shuffle(arr)
print(arr)
```

```python
import numpy as np

arr = np.array(["A", "B", "C", "D"])
np.random.shuffle(arr)
print(arr)
```

```python
import numpy as np

arr = np.array([[1, 2],
                [3, 4],
                [5, 6]])
np.random.shuffle(arr)
print(arr)
#For a 2D array, shuffle() shuffles the rows, not the individual elements.
#he values inside each row remain together.
```

# 🔷 permutation()

`permutation()` returns a randomly shuffled copy of an array.

Unlike `shuffle()`, it does **not** modify the original array.

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])
new_arr = np.random.permutation(arr)
print(new_arr)
print(arr)
```

```python
import numpy as np

arr = np.array(["A", "B", "C", "D"])
print(np.random.permutation(arr))
print(arr)
```

```python
import numpy as np

arr = np.array([[1, 2],
                [3, 4],
                [5, 6]])
print(np.random.permutation(arr))
```

# 🔷 seed()

`seed()` sets the starting point for NumPy's random number generator.

Using the same seed produces the same random numbers every time.

```python
import numpy as np
np.random.seed(10)
print(np.random.rand(3))
```

```python
import numpy as np

np.random.seed(20)
print(np.random.rand(3))
```

```python
import numpy as np

np.random.seed(10)
print(np.random.rand())
print(np.random.rand())
print(np.random.rand())
```

```python
import numpy as np

np.random.seed(10)
print(np.random.rand(3))
print(np.random.rand(3))
```
