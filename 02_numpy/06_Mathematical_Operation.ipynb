# 🔷 Arithmetic Operations (`+`, `-`, `*`, `/`)

Arithmetic operations perform element-wise calculations on NumPy arrays or between an array and a scalar.

The arrays must have compatible shapes (broadcasting rules apply).

```python
import numpy as np

arr = np.array([10, 20, 30])

print("Addition:", arr + 5)
print("Subtraction:", arr - 5)
print("Multiplication:", arr * 5)
print("Division:", arr / 5)
```

```python
import numpy as np

a = np.array([10, 20, 30])
b = np.array([1, 2, 3])

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
```

```python
import numpy as np

a = np.array([[2, 4],
              [6, 8]])

b = np.array([[1, 2],
              [3, 4]])

print("Addition:")
print(a + b)

print("\nMultiplication:")
print(a * b)
```

# 🔷 Universal Functions (ufuncs)

Universal functions (ufuncs) are built-in NumPy functions that perform operations element-wise on arrays.

They are fast, optimized, and work on arrays of any shape.

```python
import numpy as np

a = np.array([10, 20, 30])
b = np.array([1, 2, 3])

print(np.add(a, b))
print(np.multiply(a, b))
```

# 🔷 sqrt()

`sqrt()` returns the square root of each element in a NumPy array.

Negative numbers produce `nan` (Not a Number) for real-valued arrays.

```python
import numpy as np

arr = np.array([1, 4, 9, 16, 25])
print(np.sqrt(arr))
```

```python
import numpy as np

arr = np.array([0, 36, 49, 64])
print(np.sqrt(arr))
```

```python
import numpy as np

arr = np.array([4, 9, -16, 25])
print(np.sqrt(arr)) # -16 sqrt not allowed
```

# 🔷 power()

`power()` raises each element of an array to the given exponent.

It performs element-wise exponentiation and supports arrays or scalars as exponents.

```python
import numpy as np

arr = np.array([2, 3, 4])
print(np.power(arr, 2))
print(np.power(arr, 3))
```

```python
import numpy as np

base = np.array([2, 3, 4])
exponent = np.array([3, 2, 1])
print(np.power(base, exponent))
```

# 🔷 abs()

`abs()` returns the absolute (positive) value of each element in a NumPy array.

Negative numbers become positive, while positive numbers and zero remain unchanged.

```python
import numpy as np

arr = np.array([-10, -5, 0, 5, 10])
print(np.abs(arr))
```

```python
import numpy as np

arr = np.array([-2.5, 3.8, -7.1])
print(np.abs(arr))
```

```python
import numpy as np

arr = np.array([[-1, 2],
                [-3, 4]])

print(np.abs(arr))
```

# 🔷 exp()

`exp()` calculates the exponential value (eˣ) of each element in a NumPy array.

It returns \(e\) raised to the power of every element, where \(e \approx 2.71828\).

```python
import numpy as np

arr = np.array([0, 1, 2])
print(np.exp(arr))
```

```python
import numpy as np

arr = np.array([-1, 0, 1])
print(np.exp(arr))
```

```python
import numpy as np

arr = np.array([[0, 1],
                [2, 3]])

print(np.exp(arr))
```

# 🔷 log()

`log()` computes the natural logarithm (base e) of each element in a NumPy array.

The input values must be positive. `log(0)` gives `-inf` and negative values produce `nan`.

```python
import numpy as np

arr = np.array([1, np.e, np.e**2])
print(np.log(arr))
```

```python
import numpy as np

arr = np.array([1, 10, 100])
print(np.log(arr)) # It is log base e
print(np.log10(arr))
```

```python
import numpy as np

arr = np.array([1, 0, -5])
print(np.log(arr))
```

# 🔷 sin()

`sin()` computes the sine of each angle in a NumPy array.

The input angles must be in **radians**, not degrees.

```python
import numpy as np

angles = np.array([0, np.pi/2, np.pi])
result = np.sin(angles)
print(np.round(result, 2))
```

```python
import numpy as np

angles = np.array([30, 45, 90])
radians = np.deg2rad(angles)
print(np.sin(radians))
```

# 🔷 cos()

`cos()` computes the cosine of each element in a NumPy array.

The input values must be in **radians**, not degrees.

```python
import numpy as np

angles = np.array([0, np.pi/2, np.pi])
print(np.cos(angles))
```

```python
import numpy as np

angles = np.array([0, 30, 60, 90])
radians = np.deg2rad(angles)
print(np.cos(radians))
```

```python
import numpy as np

angles = np.linspace(0, np.pi, 5)
print(np.round(np.cos(angles), 2))
```

# 🔷 tan()

`tan()` computes the tangent of each element in a NumPy array.

Input must be in **radians**, not degrees.

```python
import numpy as np

angles = np.array([0, np.pi/4, np.pi/2])
print(np.tan(angles))
```

```python
import numpy as np

angles = np.array([0, 30, 45, 60, 90])
radians = np.deg2rad(angles)
print(np.round(np.tan(radians), 2))
```

```python
import numpy as np

x = np.linspace(0, np.pi/2, 5)
print(np.round(np.tan(x), 2))
```
