```python
#Creation of Dataset
import numpy as np

student_data = np.array([
    [101, 85, 78, 92, 88, 80],
    [102, 67, 74, 70, 72, 68],
    [103, 95, 91, 89, 96, 94],
    [104, 45, 52, 48, 55, 50],
    [105, 76, 81, 79, 84, 77],
    [106, 88, 90, 86, 91, 89],
    [107, 59, 63, 61, 58, 60],
    [108, 99, 98, 97, 100, 99],
    [109, 73, 70, 75, 78, 72],
    [110, 82, 85, 84, 83, 81],
    [111, 39, 42, 38, 45, 40],
    [112, 91, 89, 94, 92, 90],
    [113, 64, 66, 62, 65, 63],
    [114, 55, 58, 60, 57, 59],
    [115, 79, 76, 81, 80, 78],
    [116, 93, 95, 92, 94, 96],
    [117, 48, 50, 46, 49, 47],
    [118, 71, 73, 69, 74, 72],
    [119, 86, 84, 88, 87, 85],
    [120, 62, 60, 64, 61, 63]
])

student_data
```

```python
#Save the Dataset
np.savetxt(
    "student_marks.csv",
    student_data,
    delimiter=",",
    fmt="%d"
)

print("Dataset saved successfully!")
```

```python
#Load the Dataset
data = np.loadtxt(
    "student_marks.csv",
    delimiter=",",
    dtype=int
)

print(data)
```

```python
#Verify the Dataset
print("Shape:", data.shape)
print("Dimensions:", data.ndim)
print("Data Type:", data.dtype)
print("Total Values:", data.size)
```

Analysis

The dataset contains:

20 students,
6 columns,
2-dimensional array,
120 total values

```python
# Display the First Five Student Records
print(data[:5])
```

```python
#Extract Subject Columns
student_id = data[:, 0]
math = data[:, 1]
science = data[:, 2]
english = data[:, 3]
computer = data[:, 4]
social = data[:, 5]
print("Math Marks:")
print(math)

print("\nScience Marks:")
print(science)
```

```python
#Calculate Total 
total_marks = math + science + english + computer + social

print(total_marks)
```

```python
# Calculate Percentage
percentage = (total_marks / 500) * 100

print(np.round(percentage, 2))
```

```python
# Add Derived Columns - Total Marks   - Percentage
total_marks = total_marks.reshape(-1, 1)
percentage = percentage.reshape(-1, 1)
updated_data = np.hstack((data, total_marks, percentage))
print(updated_data)
```

```python
#Find the Topper
topper = updated_data[np.argmax(updated_data[:, 7])]
print("Topper Details:")
print(topper)

```

```python
#Find the Lowest Scorer
lowest_scorer = updated_data[np.argmin(updated_data[:, 7])]
print("Lowest Scorer Details:")
print(lowest_scorer)
```

```python
#Generate the Merit List
merit_list = updated_data[np.argsort(updated_data[:, 7])[::-1]]
print(merit_list)
```

```python
#Find Students Scoring Above 80%
high_performers = updated_data[updated_data[:, 7] > 80]
print(high_performers)
```

```python
#Find Students Scoring Below 40%
below_40 = updated_data[updated_data[:, 7] < 40]
print(below_40)
```

```python
# Find Students Failing in Any Subject
failing_students = updated_data[
    (updated_data[:, 1] < 40) |
    (updated_data[:, 2] < 40) |
    (updated_data[:, 3] < 40) |
    (updated_data[:, 4] < 40) |
    (updated_data[:, 5] < 40)
]

print(failing_students)
```

```python
#Subject-wise Highest Marks
print("Highest Marks")
print("Math     :", np.max(math))
print("Science  :", np.max(science))
print("English  :", np.max(english))
print("Computer :", np.max(computer))
print("Social   :", np.max(social))
```

```python
#Subject-wise Lowest Marks
print("Lowest Marks")
print("Math     :", np.min(math))
print("Science  :", np.min(science))
print("English  :", np.min(english))
print("Computer :", np.min(computer))
print("Social   :", np.min(social))
```

```python
#Subject-wise Statistics
subjects = {
    "Math": math,
    "Science": science,
    "English": english,
    "Computer": computer,
    "Social": social
}

for subject, marks in subjects.items():
    print(f"\n{subject}")
    print("Average :", np.mean(marks))
    print("Median  :", np.median(marks))
    print("Std Dev :", round(np.std(marks), 2))
```

```python
#Find Students Scoring Above the Class Average
class_average = np.mean(updated_data[:, 7])
print("Class Average Percentage:", round(class_average, 2))
above_average = updated_data[updated_data[:,7] > class_average]
print(above_average)
```

```python
 #Search for a Student
student = updated_data[updated_data[:, 0] == 116]
print(student)
```
